##### some extra plotting options.

```python
from matplotlib import rcParams
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.colors import Normalize

# Workaround to not get error messages nowadays?
rcParams['text.usetex'] = True


# create the custom colormap.
cdict = {'red':   ((0.0, 24/255., 24/255.),
                   (0.5, 1., 1.),
                   (1.0, 164/255., 164/255.)),
         
         'green': ((0.0, 119/255., 119/255.),
                   (0.5, 1., 1.),
                   (1.0, 38/255., 38/255.)),

         'blue':  ((0.0, 149/255., 149/255.),
                   (0.5, 1, 1),
                   (1.0, 4./255., 4./255.)),
        }
brlr = LinearSegmentedColormap('brlr', cdict)
plt.register_cmap(cmap=brlr)


class MidpointNormalize(Normalize):
    """ Creates a norm that fixes a desired value to the center point of the
        colormap.?
    """
    def __init__(self, vmin=None, vmax=None, midpoint=None, clip=False):
        self.midpoint = midpoint
        if abs(vmin) > abs(vmax):
            Normalize.__init__(self, vmin, -vmin, clip)
        else:
            Normalize.__init__(self, -vmax, vmax, clip)

    def __call__(self, value, clip=None):
        # I'm ignoring masked values and all kinds of edge cases to make a
        # simple example...
#         x, y = [self.vmin, self.midpoint, self.vmax], [0, 0.5, 1]
        x, y = [self.vmin, self.midpoint, self.vmax], [0, abs(self.vmin)/(abs(self.vmin)+abs(self.vmax)), 1]
        return np.ma.masked_array(np.interp(value, x, y))
    
ms = [
    {'c' : '#114B5F', 'fc' : '#208fb5'},
    {'c' : '#47856E', 'fc' : '#59a689'},
    {'c' : '#9ca903', 'fc' : '#cadb04'},
    {'c' : '#eca300', 'fc' : '#ffba20'},
    {'c' : '#bf3100', 'fc' : '#ff4404'}
]
```

#### Data definitions.

```python
def get_data(param, data_file):
    """ Returns the array for a path in the hdf5 file.
    """
    grp = ('/dispersion_quadratic/nx_{:d}'.format(param['nx']) +
           '/coupling_{:.7g}'.format(param['coupling']) +
           '/seed_{:d}'.format(param['seed']) +
           '/beta_{:.1f}'.format(param['beta']) +
           '/n_part_{:d}-{:d}'.format(*param['n_part']) +
           '/mass_{:.1f}-{:.1f}'.format(*param['mass']) +
           '/cl_step_{:.2f}'.format(param['cl_step']) +
           '/dtau_{:s}'.format(str(round(param['dtau'], 3)))
    )
    res = {}
    with hdf.File(data_dir + data_file, 'r') as f:
        for name, data in f[grp].items():
            res[name] = data[...]
    return res

def system_numbers(n_part, nx=64):
    return {
        'ku' : n_part[0] / 2.,
        'kd' : n_part[1] / 2.,
        'q' : abs((n_part[0] - n_part[1])/2.),
        'b_low' : -nx//2,
        'h_high' : nx//2-1,
        'p' : abs((n_part[0] - n_part[1])/(n_part[0] + n_part[1])),
        'n' : abs((n_part[0] + n_part[1])/nx)
    }
```

```python
def average_and_fft(ospdm):
    """ Produces a pair distribution from the spatially distributed on-side pair density matrix.
    """
    vol = np.shape(ospdm)[0]
    
    # Average over spatial extent.
    r_dist = np.zeros(vol)
    for r in range(vol):
        r_dist += np.concatenate((ospdm[r,r:], ospdm[r,:r])) / vol
    
    # Wrap around for fft, perform fft and return result.
    res = np.fft.fftshift(np.fft.fft(r_dist)).real
    return res

def straighten(dm):
    """ From diagonal form to straight form.
    """
    l = np.shape(dm)[0]
    res = np.zeros_like(dm)
    for r in range(l):
        res[r, :] = np.concatenate((dm[r,r:], dm[r,:r]))
    return res
```

# Figure 3:  constant particle number, different polarizations (mass balanced).

## calculate the data first.

```python
param = {
#     'coupling' : 0.40625,
    'coupling' : 0.8125,
#     'coupling' : 1.21875,
    'nx' : 64,
    'cl_step' : 0.04,
    'dtau' : 0.05,
}


seeds = [774,391,274,729,224]
couplings = [0.40625, 0.8125, 1.21875]
masses = [(1.0,1.0)]
n_parts = [(23,3), (21,5), (19,7), (17,9), (15,11), (13,13)]
betas = np.arange(2.5, 7.51, 0.5)


shot_data = np.zeros(shape=(len(n_parts), len(masses), len(betas), param['nx'], param['nx']))
for n, n_part in enumerate(n_parts):
    param['n_part'] = n_part

    for m, mass in enumerate(masses):
        param['mass'] = mass

        for l, seed in enumerate(seeds):
            param['seed'] = seed

            for k, beta in enumerate(betas):
                param['beta'] = beta
                ds = get_data(param, data_file='5_paper_production_array.hdf5')

                if (n_part[0] == n_part[1]) and (mass[0] == mass[1]):
                    shot = np.fft.fftshift(
                        ds['nn_0-0'].real - np.outer(np.diag(ds['nk_1'].real), np.diag(ds['nk_1'].real))
                    )
                else:
                    shot = np.fft.fftshift(
                        ds['nn_0-1'].real - np.outer(np.diag(ds['nk_1'].real), np.diag(ds['nk_2'].real))
                    )
                shot_data[n,m,k,:,:] += shot/len(seeds)
```

```python
m = masses.index((1.0,1.0))
lpos = (1.1, 2.5)

with plt.style.context('phd_thesis'):
    n_cols, n_rows = 3, len(n_parts)//3
    fig, axes = plt.subplots(n_rows, n_cols)
    fig.set_size_inches(6.3, 4.27)
    ax = axes.flatten()
    
    for rr, n_part in enumerate(n_parts[::-1]):
        n = n_parts.index(n_part)
        
        r = rr // n_cols
        c = rr % n_cols
        
        shot = np.mean(shot_data[n,m,-2:,:,:], axis=0)
        if n_part[0] > n_part[1]:
            shot = shot.T
        norm = MidpointNormalize(midpoint=0, vmin=np.min(shot), vmax=np.max(shot))
            
        im = ax[r*n_cols+c].imshow(shot, origin='lower',
                     extent=(-np.pi, np.pi, -np.pi, np.pi),
                     norm=norm, interpolation='none', cmap=plt.get_cmap('brlr'),
        )
        ax[r*n_cols+c].text(*lpos, '$p = {:.2g}$'.format(abs(n_part[0]-n_part[1])/(sum(n_part))))
    
    for a in ax:
        a.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
        a.set_yticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
        a.set_xticklabels([])
        a.set_yticklabels([])
        a.set_xticks([-3*np.pi/4, -np.pi/4, np.pi/4, 3*np.pi/4], minor=True)
        a.set_yticks([-3*np.pi/4, -np.pi/4, np.pi/4, 3*np.pi/4], minor=True)
                
    for r in range(n_rows):
        i = r*n_cols
        ax[i].set_yticklabels(["$-\pi$", "$-\pi/2$", "$0$", "$\pi/2$", "$\pi$"])
        ax[i].set_ylabel('$k\'$')
    for c in range(n_cols):
        i = (n_rows-1)*n_cols + c   
        ax[i].set_xlabel('$k$')
        ax[i].set_xticklabels(["$-\pi$", "$-\pi/2$", "$0$", "$\pi/2$", "$\pi$"])
        
        
    fig.subplots_adjust(left=0.09, bottom=0.12, right=0.99, top=0.985, wspace=0.04, hspace=0.06)
    fig.savefig('1d_gsp_shotnoise_polarization.pdf')
```

```python
nx = 64

# Some general plot settings.
lb = -nx/2.
ub = nx/2.-1
r = 0.5/32 * nx
lw = 1
off = nx/6.

# pcolors = {
#     'normal' : {'inner' : '#595457', 'outer' : '#343330'},
#     'light' : {'inner' : '#94A727', 'outer' : '#566C2B'},
#     'heavy' : {'inner' : '#FFA737', 'outer' : '#DC851F'},
# }
pcolors = {
    'normal' : {'inner' : '#595457', 'outer' : '#343330'},
    'light' : {'inner' : '#BD5142', 'outer' : '#af2b18'},
    'heavy' : {'inner' : '#459CAF', 'outer' : '#007791'},
}

def particle(ax, coords, color, r=r, z=10):
    """ Draws a particle at ax/coords with a given color.
    """
    ax.add_artist(plt.Circle(coords, r, color=pcolors[color]['inner'], zorder=z, fill=True, lw=lw))
    ax.add_artist(plt.Circle(coords, r, color=pcolors[color]['outer'], zorder=z+1,linewidth=3, fill=False, lw=lw))
    

def draw_system(ax, nu, nd):
    """ Draws the k-k' system.
    """
    ku, kd = nu/2, nd/2
    q = ku - kd
    
    ax.set_xlim(lb, ub)
    ax.set_ylim(lb, ub)
    ax.set_aspect(1)
    
    ax.set_xticks([lb, -ku, 0, ku, ub])
    ax.set_xticklabels(['$-\\pi$', '$-k_{F}^{\\uparrow}$', '$0$', '$k_{F}^{\\uparrow}$', '$\\pi$'])
    ax.set_xticks([], minor=True)
    ax.set_yticks([lb, -kd, 0, kd, ub])
    ax.set_yticklabels(['$-\\pi$', '$-k_{F}^{\\downarrow}$', '$0$', '$k_{F}^{\\downarrow}$', '$\\pi$'])
    ax.set_yticks([], minor=True)
    
#     ax.set_xticks([lb, 0, ub])
#     ax.set_xticklabels(['$-\\pi$', '$0$', '$\\pi$'], fontsize=16)
#     ax.set_yticks([lb, 0, ub])
#     ax.set_yticklabels(['$-\\pi$', '$0$', '$\\pi$'], fontsize=16)
    
    # grid.
    ax.axhline(0, linestyle='--', linewidth=.2, color='gray', marker='', zorder=0)
    ax.axvline(0, linestyle='--', linewidth=.2, color='gray', marker='', zorder=0)
    ax.add_artist(
        plt.Rectangle((lb, -kd), nx, 2*kd, zorder=0, alpha=0.3, color='#595457', lw=0)
    )
    ax.add_artist(
        plt.Rectangle((-ku, lb), 2*ku, nx, zorder=0, alpha=0.3, color='#595457', lw=0)
    )
    
    
    # anti-diagonals.
    ax.arrow(lb, ub-q+1, nx, -nx, linestyle='--', zorder=1, lw=0.75)
    ax.arrow(lb+q, ub+1, nx, -nx, linestyle='--', zorder=1, lw=0.75)
#     ax.arrow(lb, ub+1, nx, -nx, linewidth=0.2, linestyle='--', color='gray', zorder=0)
    
    
    # Some pairing positions.
#     particle(ax, coords=(ku, -kd), color='normal')
    particle(ax, coords=(ku-2*kd, kd), color='heavy')
#     particle(ax, coords=(-ku, 2*ku-kd), color='light')
    
    particle(ax, coords=(-ku, kd), color='normal')
#     particle(ax, coords=(-ku+2*kd, -kd), color='heavy')
#     particle(ax, coords=(ku, -2*ku+kd), color='light')

#     ax.text(ub-13, ub-4, '$P = {:.2f}$'.format(np.abs(nu-nd)/float(nu+nd)))
    
    
def _draw_kspace_setup(ax, nu, nd):
    """ Draws an empty k-space setup.
    """
    e = 5
    ax.set_xlim(lb-e, ub+e)
    ax.set_ylim(lb-e, ub+e)
    ax.set_aspect(1)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xticks([], minor=True)
    ax.set_yticks([], minor=True)
    ax.set_xlabel('')
    for spine in ax.spines.values():
        spine.set_edgecolor('white')

    ax.arrow(lb, off, nx, 0, head_length=0)
    ax.arrow(lb, -off, nx, 0, head_length=0)
    
    # Label stuff.
    ku, kd = nu/2, nd/2
    q = ku - kd
    
    # ticks.
    t = off/4.3
    
    ax.text(lb-2*t, -off-5*t, '$-\\pi$')
    ax.arrow(lb, off-t, 0, 2*t, head_length=0)
    ax.arrow(lb, -off-t, 0, 2*t, head_length=0)
    
    ax.text(-0.5*t, -off-5.5*t, '$0$')
    ax.arrow(0, off-t, 0, 2*t, head_length=0, ls='--', lw=0.35, color='gray', zorder=0)
    ax.arrow(0, -off-t, 0, 2*t, head_length=0, ls='--', lw=0.35, color='gray', zorder=0)
    
    ax.text(ub, -off-5*t, '$\\pi$')
    ax.arrow(ub+1, off-t, 0, 2*t, head_length=0)
    ax.arrow(ub+1, -off-t, 0, 2*t, head_length=0)
    
    
    # kf_up/kf_down
    ax.add_artist(
        plt.Rectangle((-ku, 3*off/4), 2*ku, off/2, zorder=0, alpha=0.3, color='#595457', lw=0)
    )
    ax.add_artist(
        plt.Rectangle((-kd, -5*off/4), 2*kd, off/2, zorder=0, alpha=0.3, color='#595457', lw=0)
    )
    
    return ku, kd, q
    
    
def draw_kspace_hm(ax, nu, nd):
    """ Puts *heavy* majority pairing particles into the k-space. 
    """
    ku, kd, q = _draw_kspace_setup(ax, nu, nd)
    particles = [
#         [(ku, off), (-kd, -off), 'normal'],
        [(-ku, off), (kd, -off), 'normal'],
        [(ku-2*kd, off), (kd, -off), 'heavy'],
#         [(-ku+2*kd, off), (-kd, -off), 'heavy_inv']
    ]
    for p_up, p_down, p_col in particles:
        ax.arrow(p_up[0], p_up[1], -p_up[0]+p_down[0], -p_up[1]+p_down[1], linestyle='dashed', color=pcolors[p_col]['outer'], ls='--', lw=0.75)
        particle(ax, coords=p_up, color=p_col, r=r)
        particle(ax, coords=p_down, color=p_col, r=0.6*r)

    
def draw_kspace_lm(ax, nu, nd):
    """ Puts *light* majority pairing particles into the k-space. 
    """
    ku, kd, q = _draw_kspace_setup(ax, nu, nd)
    particles = [
#         [(ku, off), (-kd, -off), 'normal'],
        [(-ku, off), (kd, -off), 'normal'],
        [(-ku, off), (2*ku-kd, -off), 'light'],
#         [(ku, off), (-2*ku+kd, -off), 'light_inv']
    ]
    for p_up, p_down, p_col in particles:
        ax.arrow(p_up[0], p_up[1], -p_up[0]+p_down[0], -p_up[1]+p_down[1], color=pcolors[p_col]['outer'], ls='--', lw=0.75)
        particle(ax, coords=p_up, color=p_col, r=0.6*r)
        particle(ax, coords=p_down, color=p_col, r=r)
```

```python
def get_cut_lines(masses, shot_dat, n_part=(17,9), hm=True):
    """ Returns the lines along a certain cut of the shot noise.
    """
    n = n_parts.index(n_part)
    sn = system_numbers(n_part)
    k_cm = int(sn['q'])

    # Very dirty: index correction.
    if hm:
        off = len(masses)-1
    else:
        off=0
    
    lines = []
    for m, mass in enumerate(masses):
        shot = np.mean(shot_data[n,m+off,-3:-1,:,:], axis=0)
        d = [shot[k, np.mod(-k-int(k_cm), param['nx'])] for k in range(param['nx'])]
        lines.append(d)
    return lines
```
