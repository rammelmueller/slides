## Shotnoise.

```python
def get_data(param, data_file):
    """ Returns the array for a path in the hdf5 file.
    """
    grp = ('/dispersion_quadratic/nx_{:d}'.format(param['nx']) +
           '/coupling_{:.7g}'.format(param['coupling']) +
           '/seed_{:d}'.format(param['seed']) +
           '/beta_{:.1f}'.format(param['beta']) +
           '/n_part_{:d}-{:d}'.format(*param['n_part']) +
           '/mass_{:.1f}-{:.1f}'.format(*param['mass']) +
           '/cl_step_{:.2f}'.format(param['cl_step']) +
           '/dtau_{:s}'.format(str(round(param['dtau'], 3)))
    )
    res = {}
    with hdf.File(data_dir + data_file, 'r') as f:
        for name, data in f[grp].items():
            res[name] = data[...]
    return res

def system_numbers(n_part, nx=64):
    return {
        'ku' : n_part[0] / 2.,
        'kd' : n_part[1] / 2.,
        'q' : abs((n_part[0] - n_part[1])/2.),
        'b_low' : -l//2,
        'h_high' : l//2-1,
        'p' : abs((n_part[0] - n_part[1])/(n_part[0] + n_part[1])),
        'n' : abs((n_part[0] + n_part[1])/nx)
    }

def average_and_fft(ospdm):
    """ Produces a pair distribution from the spatially distributed on-side pair density matrix.
    """
    vol = np.shape(ospdm)[0]
    
    # Average over spatial extent.
    r_dist = np.zeros(vol)
    for r in range(vol):
        r_dist += np.concatenate((ospdm[r,r:], ospdm[r,:r])) / vol
    
    # Wrap around for fft, perform fft and return result.
    res = np.fft.fftshift(np.fft.fft(r_dist)).real
    return res

def straighten(dm):
    """ From diagonal form to straight form.
    """
    l = np.shape(dm)[0]
    res = np.zeros_like(dm)
    for r in range(l):
        res[r, :] = np.concatenate((dm[r,r:], dm[r,:r]))
    return res

from matplotlib import rcParams
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.colors import Normalize

# Workaround to not get error messages nowadays?
rcParams['text.usetex'] = True


# create the custom colormap.
cdict = {'red':   ((0.0, 24/255., 24/255.),
                   (0.5, 1., 1.),
                   (1.0, 164/255., 164/255.)),
         
         'green': ((0.0, 119/255., 119/255.),
                   (0.5, 1., 1.),
                   (1.0, 38/255., 38/255.)),

         'blue':  ((0.0, 149/255., 149/255.),
                   (0.5, 1, 1),
                   (1.0, 4./255., 4./255.)),
        }
brlr = LinearSegmentedColormap('brlr', cdict)
plt.register_cmap(cmap=brlr)


class MidpointNormalize(Normalize):
    """ Creates a norm that fixes a desired value to the center point of the
        colormap.?
    """
    def __init__(self, vmin=None, vmax=None, midpoint=None, clip=False):
        self.midpoint = midpoint
        if abs(vmin) > abs(vmax):
            Normalize.__init__(self, vmin, -vmin, clip)
        else:
            Normalize.__init__(self, -vmax, vmax, clip)

    def __call__(self, value, clip=None):
        # I'm ignoring masked values and all kinds of edge cases to make a
        # simple example...
#         x, y = [self.vmin, self.midpoint, self.vmax], [0, 0.5, 1]
        x, y = [self.vmin, self.midpoint, self.vmax], [0, abs(self.vmin)/(abs(self.vmin)+abs(self.vmax)), 1]
        return np.ma.masked_array(np.interp(value, x, y))
    
ms = [
    {'c' : '#114B5F', 'fc' : '#208fb5'},
    {'c' : '#47856E', 'fc' : '#59a689'},
    {'c' : '#9ca903', 'fc' : '#cadb04'},
    {'c' : '#eca300', 'fc' : '#ffba20'},
    {'c' : '#bf3100', 'fc' : '#ff4404'}
]
```

```python
data_dir = '/home/lukas/projects/thesis_data/1d_gsp/many_body/data_production/'

param = {
#     'coupling' : 0.40625,
    'coupling' : 0.8125,
#     'coupling' : 1.21875,
    'nx' : 64,
    'cl_step' : 0.04,
    'dtau' : 0.05,
}


seeds = [774,391,274,729,224]
couplings = [0.40625, 0.8125, 1.21875]
masses = [(1.0,1.0)]
n_parts = [(23,3), (21,5), (19,7), (17,9), (15,11), (13,13)]
betas = np.arange(2.5, 7.51, 0.5)


shot_data = np.zeros(shape=(len(n_parts), len(masses), len(betas), param['nx'], param['nx']))
for n, n_part in enumerate(n_parts):
    param['n_part'] = n_part

    for m, mass in enumerate(masses):
        param['mass'] = mass

        for l, seed in enumerate(seeds):
            param['seed'] = seed

            for k, beta in enumerate(betas):
                param['beta'] = beta
                ds = get_data(param, data_file='5_paper_production_array.hdf5')

                if (n_part[0] == n_part[1]) and (mass[0] == mass[1]):
                    shot = np.fft.fftshift(
                        ds['nn_0-0'].real - np.outer(np.diag(ds['nk_1'].real), np.diag(ds['nk_1'].real))
                    )
                else:
                    shot = np.fft.fftshift(
                        ds['nn_0-1'].real - np.outer(np.diag(ds['nk_1'].real), np.diag(ds['nk_2'].real))
                    )
                shot_data[n,m,k,:,:] += shot/len(seeds)
```

## Mass imbalance.

```python
data_dir = '/home/lukas/projects/thesis_data/1d_mib/'

# Read CL data.
cl = pd.read_csv(data_dir + 'CL/1DMIB_CL_energies_thermalized.dat', delimiter="\s+")
ihmc = pd.read_csv(data_dir + "HMC/1DMIB_HMC_energies.dat", sep='\s+', header=0)


colors = {
    "1.0-1.0" : "#5f9049",
    "1.1-0.9" : "#2274a5",
    "1.2-0.8" : "#9ca903",
    "1.3-0.7" : "#d7b501",
    "1.4-0.6" : "#eca300",
    "1.5-0.5" : "#db7502",
    "1.6-0.4" : "#cc5101",
    "1.7-0.3" : "#bf3100",
    "1.8-0.2" : "#5e3a33",
    "1.9-0.1" : "#1a4156"
}

line_colors = ["#2274a5", "#8ea604", "#f5bb00", "#d76a03", "#bf3100"]

def dmToMass(dm):
    return "{:.1f}-{:.1f}".format(1.0 + dm, 1.0 - dm)
def massToDM(mass):
    return np.round(float(mass.split("-")[0]) - 1.0, 3)
```

```python
def kF(N, L, dim=3):
    """ Fermi wavevector.
    """
    density = float(N) / (L**dim)
    if dim == 1:
        return density * np.pi / 2.
    elif dim == 2:
        return np.sqrt(2.*np.pi*density)
    elif dim == 3:
        return (3.*density*np.pi**2)**1./3.

    raise NotImplementedError("Wrong dimension?")


def eFG_cont(N, L, dim=3):
    """ GS free gas energy in the continuum for a balanced system with periodic
        boundary conditions (i.e. continuum).

    """
    return float(dim)/float(dim+2.) * N * kF(N,L,dim)**2 / 2.


def _normalize(energy, norm, *args):
    """ Internally used to normalize. args contains the rest of the parameters,
        needed to compute eFG_cont(N, L, dim).
    """
    if norm is None:
        return energy
    elif norm == "efg":
        return energy / eFG_cont(*args)
    else:
        return energy / norm
    
def eFG_latt(N, L, dim=3, norm="efg"):
    """ GS energy of the free gas on the lattice for a balanced system with
        periodic boundary conditions.
    """
    if dim == 1:
        energy = 2 * 0.5*(2.*np.pi/float(L))**2 * float(N)/24. * (N**2/4. + 2 - 3*int((N/2) % 2))
    else:
        raise NotImplementedError("Wrong dimension?")
    return _normalize(energy, norm, N, L, dim)


def eFG_latt_massImbalance(dm, N, L, dim=3, norm="efg", s=1):
    """ Returns the non-interacting GS energy for a mass imbalanced Fermi gas.
        The mass imbalance is defined as

            m_up    = m_0 + dm
            m_down  = m_0 - dm
        and m0 is taken to be equal to 1.
    """
    return eFG_latt(N, L, dim=dim, norm=norm) / (1-s*dm**2)


def ptLO(gamma, N_part):
    """ First order lattice perturbation theory for mass-imbalanced fermions.
        ATTENTION: code would work for multiple flavours, math is (probably)
        not solid for multiple flavours (think about dimension).
    """
    return gamma*24./np.pi**2 * np.prod(N_part)/np.sum(N_part)**2
```

```python
from qmlib_stat import ScalarStat

# Plot settings.
massColors = {
    "1.0-1.0": "#2274a5",
    "1.1-0.9": "#5f9049",
    "1.2-0.8": "#9ca903",
    "1.3-0.7": "#d7b501",
    "1.4-0.6": "#eca300",
    "1.5-0.5": "#db7502",
    "1.6-0.4": "#cc5101",
    "1.7-0.3": "#bf3100",
    "1.8-0.2": "#5e3a33",
    "1.9-0.1": "#1a4156"
}

lstyle = {
    40: '-',
    50 : '--',
    60: '-.',
    70: ':',
    80: '-'
}

# Set constant physical and numerical parameters.
param = {
    "avgmaxdrift": 1.0,
    "clinterval": 1.0,
    "clstep": 0.01,
    "dtau": 0.05,
    "nflavor": 2,
    "nx": 40,
    "reg": 0.1
}


# Get the data and just keep the applicable sub-set.
# File format:
# - clinterval    
# - avgmaxdrift    
# - dtau    
# - npart    
# - nflavor    
# - nx    
# - mass    
# - betaef    
# - clstep    
# - reg    
# - gamma 
# - e_tot     sigma    N        
# - e_kin    sigma.1    N.1        
# - e_pot sigma.1    N.1
rawData = pd.read_csv(data_dir + 'CL/1DMIB_CL_energies_thermalized.dat', delimiter="\s+")
clData = rawData[
    (rawData["avgmaxdrift"] == param["avgmaxdrift"]) &
    (rawData["clinterval"] == param["clinterval"]) &
    (rawData["clstep"] == param["clstep"]) &
    (rawData["dtau"] == param["dtau"]) &
    (rawData["nflavor"] == param["nflavor"]) &
    (rawData["reg"] == param["reg"])
]
del rawData

# # Choose the lines on the plot.
nPart = [5, 5]
gammas = np.arange(-3.0, 3.1, 0.4)
masses = map(dmToMass, np.arange(0., 0.9, 0.1))

with plt.style.context('phd_thesis'):
    fig, ax = plt.subplots()
    fig.set_size_inches(4, 2.5)

    # Plot area.
    obs = ("e_tot", "sigma", "N")
    lines = []
    for nx in [40]:
        for mass in masses:
            results = []
            for gamma in gammas:
                allStat = ScalarStat(obs[0])
                bRows = clData[(clData['npart'] == "{:d}-{:d}".format(*nPart)) & (clData['gamma'] == round(gamma, 2)) & (clData["mass"] == mass) & (clData["nx"] == nx)]
                for i in range(len(bRows)):
                    row = bRows.iloc[[i]]
                    val, sig, n = float(row[obs[0]]), float(row[obs[1]] * np.sqrt(row[obs[2]])), int(row[obs[2]])
                    allStat.addValue(obs[0], mu=val, var=sig**2 * n, n=n)

                res = allStat.evaluateVariables(obs[0])
                if round(res[obs[0]][0], 2) != 0.00:
                    results.append([-gamma, res[obs[0]][0], res[obs[0]][1] / np.sqrt(res[obs[0]][2])])

            e_mbar= eFG_latt_massImbalance(massToDM(mass), np.sum(nPart), param["nx"], dim=1, norm="efg")
            if len(results):
                lab = "$\\bar m = {:.1f}$".format(massToDM(mass), nx)
                x, y, sig = zip(*results)
#                 line = ax.plot(x, y, label=lab, color=massColors[mass], linestyle=lstyle[nx], marker="")[0]
                line = ax.errorbar(x, y, yerr=sig, 
                                   color=massColors[mass], marker="s", markersize=2, label=lab)
                lines.append(line)
                if massToDM(mass) > 0.4:
                    fineAx = np.linspace(-2.5, 2.0, 100)
                    ax.plot(fineAx, np.array([e_mbar]*len(fineAx)) + ptLO(fineAx, nPart), linestyle=":", marker="", color=line[0].get_color())
                ax.plot(0., e_mbar, marker="x", markersize=4, color=massColors[mass])

                
    # -----------------------------------
    # Plot cosmetics.

#     ax.set_xlim([-0.2, 3.4])
    ax.set_xlim([-3.4, 3.4])
    ax.set_xticks([-3, -2, -1, 0, 1, 2, 3])
    ax.set_xticklabels(['$-3$', '$-2$', '$-1$', '$0$', '$1$', '$2$', '$3$'])
    ax.set_xticks([], minor=True)
    ax.set_xlabel("$\\gamma$")
    
    ax.set_ylabel("$E/E_{\\mathrm{FG}}}$")
    ax.set_yticks([-2, -1, 0, 1, 2, 3, 4])
    ax.set_yticklabels(['$-2$', '$-1$', '$0$', '$1$', '$2$', '$3$', '$4$'])
    ax.set_yticks([], minor=True)

    ax.axvline(ymin=-10, ymax=10, x=0, linestyle=":",color="gray", linewidth=0.5)
    ax.text(-3.1, 3.5, "{:d}$\\uparrow$ + {:d}$\\downarrow$".format(*nPart))

    ax.legend(loc="lower right", ncol=3, handles=lines)
    fig.subplots_adjust(left=0.1, bottom=0.15, right=0.98, top=0.99)
    fig.savefig("mib_eos.svg", facecolor=bg_color)
```
