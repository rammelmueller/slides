### Useful definitions.

```python
from scipy.optimize import curve_fit
from scipy.interpolate import interp1d
from scipy.interpolate import CubicSpline

# Import the module for the spline interpolation + uncertainties.
import sys
sys.path.append('/home/lukas/Dropbox/hpc_python/interpolation_uncertainty/')
from cs_uncert import CubicSplineUncertainty


def spline_line(x_data, y_data, sig_data=None, npoints=200, order=0):
    """ Takes x and y values and returns a fine spline.
    """
    xf = np.linspace(np.min(x_data), np.max(x_data), npoints)
    if sig_data is not None:
        csu = CubicSplineUncertainty(x_data, y_data, sig_data)
        yf, sigf = csu.interpolate(xf, order=order)
        return xf, yf, sigf
    else:
        cs = CubicSpline(x_data, y_data, bc_type='natural')
        yf = cs(xf) if not order else cs(xf, order)
        return xf, yf
    

def discretize_line(x_fine, y_fine, x_grid, sig_fine=None, kind='cubic'):
    """ Interpolates a fine-grained function to some specific values.
    """
    if sig_fine is not None:
        csu = CubicSplineUncertainty(x_fine, y_fine, sig_fine)
        return csu.interpolate(x_grid, order=0)
    else:    
        cs = CubicSpline(x_fine, y_fine, bc_type='natural')
        return cs(x_grid)

def n_derive(x_data, y_data, order=1):
    """ N-th order numerical derivative.
    """
    d = np.zeros_like(y_data)
    d[:-1] = np.diff(y_data) / np.diff(x_data)
    d[-1] = (y_data[-1] - y_data[-2])/(x_data[-1] - x_data[-2])
    return d


def get_full_line(param, x, y, warmup_rate=0.2, frequency=1, x_max=2.5, x_min=-2.0, norm='n0'):
    """ Returns x and y values according to the specified parameters
        and averages over all available seeds.
    """
    
    if norm is not None:
        ds = cl.loc[(cl[list(param)] == pd.Series(param)).all(axis=1)][[x, 'n_up', 'n_down', 'seed', norm]]
        if y == 'n':
            ds[y] = (cl['n_up'] + cl['n_down']) / cl[norm] 
        if y == 'm':
            ds[y] = (cl['n_up'] - cl['n_down']) / cl[norm] 
    else:
        ds = cl.loc[(cl[list(param)] == pd.Series(param)).all(axis=1)][[x, 'n_up', 'n_down', 'seed']]
        if y == 'n':
            ds[y] = (cl['n_up'] + cl['n_down']) / cl['nx']**3
        if y == 'n_squ':
            ds[y] = (cl['n_up'] + cl['n_down'])**2 / cl['nx']**6
            print('test')
        if y == 'm':
            ds[y] = (cl['n_up'] - cl['n_down']) / cl['nx']**3
        if y == 'm_squ':
            ds[y] = (cl['n_up'] - cl['n_down'])**2 / cl['nx']**6
    
    ds[y] = ds[y].map(lambda x: x[int(warmup_rate*len(x))::frequency])
    ds['samples'] = ds[y].map(len)
    ds['avg'] = ds[y].map(np.mean)
    ds['std'] = ds[y].map(np.std)
    ds['std'] /= np.sqrt(ds['samples'])
    
    ds['avg_n'] = ds['avg'] * ds['samples']
    ds['std2_n2'] = ds['std']**2 * ds['samples']**2
    
    # Group for every datapoint and calculate the averages over different runs.
    grps = ds.groupby([x], as_index=False)
    ds = grps.agg({'avg_n':'sum', 'std2_n2':'sum', 'samples':'sum'})
    ds['avg'] = ds['avg_n'] / ds['samples']
    ds['std'] = np.sqrt(ds['std2_n2'] / ds['samples']**2)
    
    return ds[(ds[x]<=x_max) & (ds[x]>=x_min)][[x, 'avg', 'std']]


def tcl_extrapolated_line(param, x, y, atols=[0.04, 0.08, 0.12, 0.16], dx=0.5, **kwargs):
    """ Extrapolates a line as a function of beta*mu to zero adaptive tolerance and returns
        a value of points with errors.
    """
    # First get the right data.
    bmus = np.arange(kwargs['x_min'], kwargs['x_max']+0.1, dx)
    ex_data = np.zeros(shape=(len(atols),len(bmus),2))
    for k, atol in enumerate(atols):
        param['adaptiveTol'] = atol
        ds = get_full_line(param, x, y, **kwargs)
        ex_data[k,:,0], ex_data[k,:,1] = ds['avg'].real, ds['std']

    # Extrapolate to zero tolerance.
    extrap = np.zeros(shape=(len(bmus), 3))
    efunc = lambda x, a, b: a*x + b
    for b, bmu in enumerate(bmus):
        popt, pcov = curve_fit(efunc, atols, ex_data[:,b,0], sigma=ex_data[:,b,1])
        extrap[b,0], extrap[b,1], extrap[b,2] = bmu, popt[1], np.sqrt(pcov[-1,-1])
        
    return extrap[:,0], extrap[:,1], extrap[:,2]
```

## Density EOS (balanced)

```python
import matplotlib.patches as patches

param = {
    'nx': 11,
    'beta': 8.0,
    'CLTau' : 1.0,
    'h' : 0.0,
}
bmu_min, bmu_max = -3., 2.5
npoints = 500
ve_min = -10.

with plt.style.context('phd_thesis'):
    fig, ax = plt.subplots()
    fig.set_size_inches(4, 2.5)
    
    lines = []
    
    
    # Density (upper panel)
    # ------------------------------------------------------------------------------------------------------------------
    
    # CL
    x, y, sig = tcl_extrapolated_line(param, 'mu', 'n', x_min=-3., x_max=2.5)
    line = ax.errorbar(x, y, yerr=sig, color=ls[0]['c'], markerfacecolor=ls[0]['fc'], label='CL'.format(param['h']), zorder=10, marker='s')
    lines.append(line)
      
    # Drut et al. DHMC data, 2012.
    x, y, sig = compare['dhmc_drut_density']['bmu'], compare['dhmc_drut_density']['nn0'], compare['dhmc_drut_density']['sig']
    m = x <= bmu_max
    line = ax.errorbar(x[m], y[m], yerr=sig[m], label='DHMC', marker=cstyle['dhmc']['m'], color=cstyle['dhmc']['c'], markerfacecolor=cstyle['dhmc']['fc'], zorder=2)
    lines.append(line)
    
    # BDMC data.
    x, y, sig = compare['bdmc_density']['bmu'], compare['bdmc_density']['nn0'], compare['bdmc_density']['sig']
    line = ax.errorbar(x, y, yerr=sig, label='BDMC', marker=cstyle['bdmc']['m'], color=cstyle['bdmc']['c'], markerfacecolor=cstyle['bdmc']['fc'], zorder=3)
    lines.append(line)

    # Frank, Luttinger Ward.
    slwf = compare['lw_frank_thermo'][compare['lw_frank_thermo']['bh']==param['h']]
    line = ax.plot(slwf['bmu'], slwf['nn0_pol'], label='Luttinger Ward', 
            marker=cstyle['lw']['m'], color=cstyle['lw']['c'], zorder=1, markerfacecolor=cstyle['lw']['fc'])
    lines.append(line[0])
    
    # Zwierlein experimental data, 2012.
    x, y, sig_y = compare['mit_nn0_bmu']['bmu'], compare['mit_nn0_bmu']['nn0'], compare['mit_nn0_bmu']['sig_nn0']
    m = x <= 3.
    line = ax.errorbar(x[m], y[m], yerr=sig_y[m], label='experiment (MIT)', marker=cstyle['mit']['m'], color=cstyle['mit']['c'], markerfacecolor=cstyle['mit']['fc'], zorder=1, markersize=2)
    lines.append(line)
    
    # Virial expansion.
    ve_mus = np.linspace(-4, -0.25, 50) / param['beta']
    ve = ve_density(ve_mus, param['h']/param['beta'], param['beta'], param['nx'], norm='mean')
    line = ax.plot(ve_mus*param['beta'], ve, label='3rd order virial exp.', linestyle='--', marker='', color=cstyle['ve']['c'])
    lines.append(line[0])
    
    
    # PT area.
    ax.fill_between([2.5, 3.2], [0, 0], [4, 4], facecolor='#dddddd', linewidth=0)
    
    ax.set_xlabel('$\\beta\\mu$', labelpad=2)
    ax.set_xlim(-4., 3.)
    ax.set_xticks([-4, -3, -2, -1, 0, 1, 2, 3])
    ax.set_xticks([-3.5, -2.5, -1.5, -0.5, 0.5, 1.5, 2.5], minor=True)
    ax.set_xticklabels(['$-4$', '$-3$', '$-2$', '$-1$', '$0$', '$1$', '$2$', '$3$'])

    ax.set_ylabel('$n/n_0$', labelpad=2)
    ax.set_ylim(0.9, 3.6)
    ax.set_yticks([1, 1.5, 2, 2.5, 3, 3.5])
    ax.set_yticks([1.25, 1.75, 2.25, 2.75, 3.25], minor=True)
    ax.set_yticklabels(['$1$', '$1.5$', '$2$', '$2.5$', '$3$', '$3.5$'])
    
    ax.legend(loc='upper left', handles=lines) 
    
    # Export.
    rect = patches.Rectangle((-100,-100), 200, 200,linewidth=0,facecolor='white', zorder=0)
    ax.add_patch(rect)
    
    fig.subplots_adjust(left=0.1, bottom=0.12, right=0.975, top=0.98)
    fig.savefig('UFG_density_balanced.svg', transparent=True)
```

```python
from copy import copy

def get_chi_spline(arg_param, bmus=[], bh_min=0.4, bh_max=2.1, dbh=0.4):
    """ Extract the spin susceptibility via numerical derivative.
    """
    param = copy(arg_param)
    
    # CL derivatives.
    bhs = np.arange(0, bh_max, dbh)
    
    chi_m = np.zeros(shape=(len(bmus),len(bhs),2))
    for b, bmu in enumerate(bmus):
        param['mu'] = np.round(bmu, 2)

        # CL susceptibility & add the 0 at the origin.
        bhs, m, sig = tcl_extrapolated_line(param, 'h', 'm', x_min=bh_min, x_max=bh_max, dx=dbh, norm=None)
        bhs = np.concatenate(([0], bhs)) 
        m = np.concatenate(([0], m))
        sig = np.concatenate(([0], sig))

        csu = CubicSplineUncertainty(bhs, m, sig)
        bhs_spline = np.linspace(np.min(bhs), np.max(bhs), 100)
        
        dm_dbh, sigp_spline = csu.interpolate(bhs, order=1)
        
        for h in bhs:
            chi_m[b,:,0] = dm_dbh
            chi_m[b,:,1] = sigp_spline
        
    return chi_m

param = {
    'nx': 11,
    'beta': 8.0,
    'CLTau' : 1.0
}
bmus = np.arange(-3., 3.1, 0.5)
results = get_chi_spline(param, bmus=bmus)
```
