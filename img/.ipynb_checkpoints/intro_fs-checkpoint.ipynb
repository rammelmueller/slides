# Fermi surfaces.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib.patches as mpatches
```

```python
edos = lambda e, m, mu : (1./(4*np.pi**2) * (2*m)**(3/2) * np.sqrt(e))

def get_lines(mu, m, c=100):
    e_up, e_down = np.linspace(0, c+mu[0], 100), np.linspace(0, c-mu[1], 100)
    return{
        'up' : {'e' : e_up, 'd': edos(e_up, m[0], mu[0])},
        'down' : {'e' : e_down, 'd': edos(e_down, m[1], mu[1])}
    }

col = ['#af2b18', '#007791']

with plt.style.context('phd_thesis'):
    fig, ax = plt.subplots(1,3)
    fig.set_size_inches(6.3, 1.85)
    
    sets = [
        [(0,0), (1,1)],
        [(20,20), (1,1)],
        [(0,0), (1,2)],
    ]
    spread = [] 
    for k, s in enumerate(sets):
        lines = get_lines(s[0], s[1])
    
        # Down (right).
        ax[k].fill_between(lines['up']['d'], lines['up']['e'], [lines['up']['e'][-1]]*100, color=col[0], alpha=0.2, lw=0)
        ax[k].plot(lines['up']['d'], lines['up']['e'], marker='', ls='-', lw=2, color=col[0])

        # Down (left).
        ax[k].fill_between(-lines['down']['d'], lines['down']['e'], [lines['down']['e'][-1]]*100, color=col[1], alpha=0.2, lw=0)
        ax[k].plot(-lines['down']['d'], lines['down']['e'], marker='', ls='-', lw=2, color=col[1])
        
        spread.append([-np.max(lines['down']['d'])-0.1, np.max(lines['up']['d'])+0.1])
        ax[k].set_xlim(spread[-1][0], spread[-1][1])
        
    for k, a in enumerate(ax.flatten()):
        a.spines['left'].set_position('zero')
        a.spines['bottom'].set_position('zero')

        # Eliminate upper and right axes
        a.spines['right'].set_color('none')
        a.spines['top'].set_color('none')
        a.spines['left'].set_color('none')
        a.spines['bottom'].set_color('none')
        a.set_xticks([])
        a.set_yticks([])
        a.set_ylim(0, 150)
        
        mup, mdown = sets[k][0]
        a.plot([-10, 0], [100-mdown]*2, ls='--', color=col[1], lw=1, marker='')
        a.plot([0, 10], [100+mup]*2, ls='--', color=col[0], lw=1, marker='')
        
        # y-axis
        dx = spread[k][1] - spread[k][0]
        a.arrow(0,0, 0,130, 
                head_width=0.03*dx, head_length=4, overhang = 0, 
                color='black', zorder=10,
                length_includes_head= True, clip_on=False)
        
        # x-axis
        a.arrow(spread[k][0],0, dx, 0, 
                head_width=5, head_length=0.03*dx, overhang = 0, 
                color='black', zorder=10,
                length_includes_head= True, clip_on=False)
        
        off_low = .06 if k==2 else 0.06 if k==1 else 0
        off_up = .06 if k==2 else 0.0 if k==1 else 0
        a.text(0.07+off_up, 127, '$\\varepsilon$')
        a.text(0.88+off_low, 0.1, '$\\rho(\\varepsilon)$')
    
    # Magnetic split label.
    ax[1].plot([-10, 0], [100+sets[1][0][1]]*2, ls='--', marker='', color='gray')
    ax[1].plot([-0.02, 0.02], [100]*2, ls='-', color='black', lw=1, marker='')

    px, py, dy = -0.5, 100, 18
    ax[1].arrow(px, py, 0, dy, head_width=0.05, head_length=5, linewidth=1, color='gray', length_includes_head=True, zorder=10)
    ax[1].arrow(px, py, 0, -dy, head_width=0.05, head_length=5, linewidth=1, color='gray', length_includes_head=True, zorder=10)
    ax[1].text(px+0.05, py-5, '$2h$', color='gray')
    ax[1].text(0.05, 97, '$\\mu$', color='black')
    
    ax[1].text(-0.87, 77, '$\\mu_\\downarrow$', color=col[1])
    ax[1].text(0.92, 117, '$\\mu_\\uparrow$', color=col[0])
    
    
#     fig.text(0.03, 0.835, '$\\mu$', color='gray')
    fig.text(0.14, 0.03, 'balanced')
    fig.text(0.46, 0.03, '$\\mu_\\uparrow \\neq \\mu_\\downarrow$')
    fig.text(0.8, 0.03, '$m_\\uparrow \\neq m_\\downarrow$')
    
    # Arrows.
    arrow = mpatches.FancyArrowPatch((-0.25, 80), (-0.25, 40), mutation_scale=20, color=col[1], alpha=0.4, lw=0)
    ax[0].add_patch(arrow)
    arrow = mpatches.FancyArrowPatch((0.25, 40), (0.25, 80), mutation_scale=20, color=col[0], alpha=0.4, lw=0)
    ax[0].add_patch(arrow)

        
    fig.subplots_adjust(left=0.05, right=0.95, bottom=0.2)
    fig.savefig('dos.pdf')
```

### Feshbach resonance.

```python
colors = ['#af2b18', '#007791']

vdw = lambda r, a, b : 4*b*((a/r)**12 - (a/r)**6)
resonance = lambda b, b0, delta, a0 : a0*(1 - delta/(b - b0))

with plt.style.context('phd_thesis'):
    fig, ax = plt.subplots(1,2)
    fig.set_size_inches(6.3, 1.8)

    a, b = 1, 4
    dE = 2
    r = np.linspace(0.5, 5, 1000)
    line = ax[0].plot(r, vdw(r, a, b), ls='-', marker='', color=colors[1])
    ax[0].axhline(0, color=line[0].get_color(), marker='', ls='--', lw=0.5)
    ax[0].text(2.15, -0.68, '$\\mathrm{open\ channel}$', fontsize=7, color=line[0].get_color())
    
    line = ax[0].plot(r, vdw(r, a+0.04, b-0.2) + dE, ls='-', marker='', color=colors[0])
    ax[0].text(2.115, dE+0.15, '$\\mathrm{closed\ channel}$', fontsize=7, color=line[0].get_color())
    
    # Resonance line.
    res = dE/2
    ax[0].plot([1.055, 1.62], [res]*2, ls='-', color='gray', lw=0.5, marker='', zorder=1)
    ax[0].plot([1.62, 1.9], [res]*2, ls='--', color='gray', lw=0.5, marker='', zorder=1)
    
#     ax[0].annotate('', (1.8, 0), (1.8, res), arrowprops={'arrowstyle': '<->', 'color' : 'gray', 'linewidth' : 0.5, 'shrink' : 0}, va='center')
    ax[0].arrow(1.8, 0.5, 0, res-0.55, head_width=0.02, head_length=0.15, linewidth=0.5, color='gray', length_includes_head=True, zorder=10)
    ax[0].arrow(1.8, 0.5, 0, -0.45, head_width=0.02, head_length=0.15, linewidth=0.5, color='gray', length_includes_head=True, zorder=10)
    ax[0].text(1.83, res/2-0.17, '$\\Delta\\mu B$', fontsize=7, color='gray')

        
    ax[0].set_xlim(0.8,2.5)
    ax[0].set_xticks([])
    ax[0].set_xticklabels([])
    ax[0].set_xlabel('$\\mathrm{spatial\ separation}\ r$')
    
    ax[0].set_ylim(-b-1, 4)
    ax[0].set_yticks([])
    ax[0].set_ylabel('$\\mathrm{energy}\ E$')
    
    # -----

    b0, delta, a0 = 0, 0.2, 1
    b_low = np.linspace(-2, b0, 1000)
    b_high = np.linspace(b0, 2, 1000)
    
    # Lines + points.
    ax[1].plot(b_low, resonance(b_low, b0, delta, a0), ls='-', marker='', color=colors[0])
    ax[1].plot(b_high, resonance(b_high, b0, delta, a0), ls='-', marker='', color=colors[0])
    ax[1].axhline(1, ls='--', color=colors[0], marker='', lw=0.5)
    
    
    # Noninteracting point.
    pf = b0 + delta
    ax[1].plot(pf, resonance(pf, b0, delta, a0), color='gray', marker='o')
    line = ax[1].axhline(0, ls=':', color='gray', marker='', lw=0.5)
    ax[1].text(pf+0.1, resonance(pf, b0, delta, a0)-0.3, '$\\mathrm{noninteracting}$', fontsize=7, color=line.get_color())

    # BCS arrow.
    ax[1].annotate('', (1.2, 0.75), (2, 0.8), arrowprops={'arrowstyle': '<-', 'color' : '#000000', 'linewidth' : 0.75}, va='center')
    ax[1].text(1.4, 0.5, '$BCS$', fontsize=7)

    # BEC arrow.
    ax[1].annotate('', (-1.2, 1.3), (-2, 1.25), arrowprops={'arrowstyle': '<-', 'color' : '#000000', 'linewidth' : 0.75}, va='center')
    ax[1].text(-1.7, 1.35, '$BEC$', fontsize=7)
    
    # UFG point.
#     line = ax[1].plot(0, 1, color=colors[1], marker='o')
    line = ax[1].axvline(b0, ls='--', color=colors[1], marker='', lw=0.5)
    ax[1].text(0.05, 1.5, '$\\mathrm{unitary\ regime}$', fontsize=7, color=line.get_color())
    
    ax[1].set_xlim(-2, 2)
    ax[1].set_xticks([])
    ax[1].set_xlabel('$\\mathrm{magnetic\ field}\ B$')
    
    ax[1].set_ylabel('$\\mathrm{scattering\ length}\ a$')
    ax[1].set_ylim(-1*a0, 3*a0)
    ax[1].set_yticks([], minor=True)
    ax[1].set_yticks([0, 1])
    ax[1].set_yticklabels(['$0$', '$a_{\\mathrm{BG}}$'])
    
    # Change ytick colors.
    plt.gca().get_yticklabels()[0].set_color('gray')
    plt.gca().get_yticklabels()[1].set_color(colors[0])
    
    fig.subplots_adjust(left=0.05, right=0.99, bottom=0.1, wspace=0.3)
    fig.savefig('feshbach.pdf') 
```

## Pairing mechanism of two arbitrary Fermi surfaces.

```python

with plt.style.context('seaborn'):
    fig, ax = plt.subplots()
    fig.set_size_inches(10,10)
    
    k1, k2 = 10, 8
    q = 2
    dk = 0.5
    
    # draw the circles.
    theta = np.linspace(0, 2*np.pi, 100)
    
    x1, y1 = k1*np.cos(theta), k1*np.sin(theta)
    x1o, y1o = (k1+dk)*np.cos(theta), (k1+dk)*np.sin(theta)
    line = ax.plot(x1, y1)
    ax.plot(x1o, y1o, color=line[0].get_color(), ls='--', alpha=0.5)
    
    
    x2, y2 = k2*np.cos(theta)+q, k2*np.sin(theta)
    x2o, y2o = (k2+dk)*np.cos(theta)+q, (k2+dk)*np.sin(theta)
    line = ax.plot(x2, y2)
    ax.plot(x2o, y2o, color=line[0].get_color(), ls='--', alpha=0.5)
#     ax.fill_between(x2o, y2, y2o)

    # Anchor points.
    aw, ah = 0.4, 0.4
    ax.plot(0,0, marker='o', color='gray')
#     ax.plot(q,0, marker='o', color='black')
    ax.arrow(0, 0, q, 0, head_width=aw, head_length=ah, linewidth=1, color='gray', length_includes_head=True, zorder=10)
      
    # Cosmetics.
    ax.set_aspect(1)
```

```python
from matplotlib.patches import Circle, FancyArrowPatch
from matplotlib.collections import PatchCollection

def plot_fs(ax, k, dk=2, shift=0, color='black', zorder=5):
    """ Draws a Fermi surface + scattering range.
    """
#     ax.plot(x, y, ls='-', marker='', color=color)
#     ax.plot(xo, yo, color=color, ls='--', alpha=0.5, marker='')
    
    patches = []
    patches.append(Circle((shift, 0), radius=k+dk))
    ax.add_collection(PatchCollection(patches, alpha=0.4, color=color, lw=0, zorder=zorder-2))
    
    patches = []
    patches.append(Circle((shift, 0), radius=k))
    ax.add_collection(PatchCollection(patches, alpha=1, color='white', lw=0, zorder=zorder-1))
    
    theta = np.linspace(0, 2*np.pi, 100)
    x, y = k*np.cos(theta)+shift, k*np.sin(theta)
    xo, yo = (k+dk)*np.cos(theta)+shift, (k+dk)*np.sin(theta)
    ax.plot(x, y, ls='-', marker='', color=color, zorder=zorder)
    

col = ['#af2b18', '#007791']

with plt.style.context('phd_thesis'):
    fig, ax = plt.subplots(1,3)
    fig.set_size_inches(6.3,2.15)
    
    k1, k2, dk = 25, 15, 1.5
    q = k1 - k2
    
    # For the arrows.
    aw, ah = 0.75, 0.75
    lw, off = 1.5, 0.75
    

    # -----------------------------------------------------------------
    # BALANCED.
    plot_fs(ax[0], k1, color='black')    

    # Initial.
    a = 45
    ax[0].arrow(0, 0, (k1-off)*np.cos(a*np.pi/180), (k1-off)*np.sin(a*np.pi/180), ls='-',
                head_width=aw, head_length=ah, linewidth=lw, color='black', length_includes_head=True, zorder=20)
    a += 180
    ax[0].arrow(0, 0, (k1-off)*np.cos(a*np.pi/180), (k1-off)*np.sin(a*np.pi/180), ls='-',
                head_width=aw, head_length=ah, linewidth=lw, color='black', length_includes_head=True, zorder=20)
    ax[0].text(10, 15, '$\\vec k_\\uparrow$', zorder=100)
    ax[0].text(-14, -18.5, '$\\vec k_\\downarrow$', zorder=100)
    
    # ---
    
    # Scattering.
    b = 5
    ax[0].arrow(0, 0, (k1-off)*np.cos(b*np.pi/180), (k1-off)*np.sin(b*np.pi/180), ls='-.',
                head_width=aw, head_length=ah, linewidth=1, color='black', length_includes_head=True, zorder=20)
    
    c = FancyArrowPatch(
        ((k1+2*dk)*np.cos((a-180)*np.pi/180), (k1+2*dk)*np.sin((a-180)*np.pi/180)), 
        ((k1+2*dk)*np.cos(b*np.pi/180), (k1+2*dk)*np.sin(b*np.pi/180)),
        connectionstyle="arc3,rad=-0.2", mutation_scale=2
    )
    ax[0].add_collection(PatchCollection([c], alpha=1, color='black', lw=0.5, zorder=100))
    
    b += 180
    ax[0].arrow(0, 0, (k1-off)*np.cos(b*np.pi/180), (k1-off)*np.sin(b*np.pi/180), ls='-.',
                head_width=aw, head_length=ah, linewidth=1, color='black', length_includes_head=True, zorder=20)
    
    c = FancyArrowPatch(
        ((k1+2*dk)*np.cos((a)*np.pi/180), (k1+2*dk)*np.sin((a)*np.pi/180)), 
        ((k1+2*dk)*np.cos(b*np.pi/180), (k1+2*dk)*np.sin(b*np.pi/180)),
        connectionstyle="arc3,rad=-0.2", mutation_scale=2,
    )
    ax[0].add_collection(PatchCollection([c], alpha=1, color='black', lw=0.5, zorder=100))

                             
    ax[0].text(20, -3, '$\\vec{k}\,\'_\\uparrow$', zorder=100)
    ax[0].text(-23, 0, '$\\vec{k}\,\'_\\downarrow$', zorder=100)
    
    
    # -----------------------------------------------------------------
    # IMBALANCED.
    plot_fs(ax[1], k1, color=col[0])
    plot_fs(ax[1], k2, color=col[1], shift=0, zorder=10)
    a = 45
    ax[1].arrow(0, 0, (k1-off)*np.cos(a*np.pi/180), (k1-off)*np.sin(a*np.pi/180), ls='-',
                head_width=aw, head_length=ah, linewidth=lw, color=col[0], length_includes_head=True, zorder=20)
    
    a += 180
    ax[1].arrow(0, 0, (k2-off)*np.cos(a*np.pi/180), (k2-off)*np.sin(a*np.pi/180), ls='-',
                head_width=aw, head_length=ah, linewidth=lw, color=col[1], length_includes_head=True, zorder=20)
     
    ax[1].text(10, 15, '$\\vec k_\\uparrow$', zorder=100, color=col[0])
    ax[1].text(-7, -11, '$\\vec k_\\downarrow$', zorder=100, color=col[1])      
    
    
    ax[1].arrow(0, -k2-q/2, 0, q/2-off, head_width=aw, head_length=ah, linewidth=lw, color='gray', length_includes_head=True, zorder=100)
    ax[1].arrow(0, -k2-q/2, 0, -q/2+off, head_width=aw, head_length=ah, linewidth=lw, color='gray', length_includes_head=True, zorder=100)
#     ax[1].text(2, -k2-q/2, '$|\\vec k_{\\mathrm{F}}^{\\uparrow}-\\vec k_{\\mathrm{F}}^{\\downarrow}|$', zorder=100, color='gray')
    ax[1].text(2, -k2-q/2-1, '$|\\vec q\,|$', zorder=100, color='gray')
    
    # -----------------------------------------------------------------
    # IMBALANCED, SHIFTED..
    plot_fs(ax[2], k1, color=col[0])
    plot_fs(ax[2], k2, color=col[1], shift=q, zorder=10)
    
    ax[2].text(q/2-1, -4, '$\\vec{q}$', zorder=100)
    ax[2].arrow(0, 0, q-0.5, 0, head_width=aw, head_length=ah, linewidth=1, color='black', length_includes_head=True, zorder=100)
    ax[2].plot(q,0, marker='o', color='gray', markersize=3, zorder=90)
    
#     a = 45
#     ax[2].arrow(0, 0, (k1-off)*np.cos(a*np.pi/180), (k1-off)*np.sin(a*np.pi/180), ls='-',
#                 head_width=aw, head_length=ah, linewidth=lw, color=col[0], length_includes_head=True, zorder=20)
#     a += 180
#     ax[2].arrow(0, 0, (k2-off)*np.cos(a*np.pi/180), (k2-off)*np.sin(a*np.pi/180), ls='-',
#                 head_width=aw, head_length=ah, linewidth=lw, color=col[1], length_includes_head=True, zorder=20)
    # -----------------------------------------------------------------    
    # Cosmetics.
    for a in ax.flatten():
        a.set_xticks([])
        a.set_yticks([])
        a.set_aspect(1)
        spread = 1.2
        a.set_xlim(-k1*spread, k1*spread)
        a.set_ylim(-k1*spread, k1*spread)
        a.axis('off')
        a.plot(0,0, marker='o', color='black', markersize=3, zorder=100)
        
    fig.subplots_adjust(left=0.01, right=0.99, wspace=0.02, hspace=0.02, top=0.99, bottom=0.01)
    fig.savefig('fflo.pdf')
```
