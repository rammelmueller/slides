```python
import numpy as np
import pandas as pd
import h5py as hdf
import matplotlib.pyplot as plt
from toadlib_dev.evaluation import DataConverter
from pylab import cm
import scipy.integrate as integrate

# Workaround to not get error messages nowadays?
from matplotlib import rcParams
rcParams['text.usetex'] = True

%matplotlib inline


# Comparison data.
prefix = '/home/lukas/Dropbox/ComplexLangevinPolarized3D/'
dmwl_data = pd.read_csv(prefix + 'Data/interpolationDens_HMC_DLWM.txt', sep='\s+')
zdata = pd.read_csv(prefix + 'Data/ndata_Zwierlein.dat', sep='\s+', header=None)
zdata_mu = pd.read_csv(prefix + 'Data/mudata_Zwierlein.dat', sep='\s+', header=None)
bdmc_data = pd.read_csv(prefix + 'Data/n_over_n0_bold_MC_17_5.dat', sep='\s+', names=['betamu', 'nn0', 'sig'], header=0)

alpha_level = 0.4

cols = {
    0.0 : '#187795',
    0.4: '#5f9049',
    0.8: '#9ca903',
    1.2: '#d7b501',
    1.6: '#eca300',
    2.0: '#db7502' 
}

def get_data(param, var, warmup_rate=0., frequency=1):
    """ Returns the array for a path in the hdf5 file.
    """ 
    data_file = prefix + 'Data/3DCLP_densities_nx{:d}_beta{:d}.hdf5'.format(param['nx'], int(param['beta']))
    grp = ('/xi_{:.1f}/'.format(param['xi']) +
           'nx_{:d}/'.format(param['nx']) +
           'beta_{:.1f}/'.format(param['beta']) +
           'h_{:.1f}/'.format(param['h']) +
            ('mu_{:.1f}/'.format(param['mu']) 
                if not (int(param['mu'])-param['mu']) else 
            'mu_{:.3g}/'.format(param['mu'])) +
           'seed_{:d}/'.format(param['seed'])
          )
    with hdf.File(data_file, 'r') as f:
        data = f[grp + var]
        cutoff = int(warmup_rate*len(data))
        return data[cutoff::frequency]


def get_full_data(param, var, warmup_rate=0., frequency=1, seeds=[4312, 5432, 6543]):
    """ `Gets a dataset across all seeds.
    """
    all_data = []
    for seed in seeds:
        param['seed'] = seed
        all_data.extend(get_data(param, var, warmup_rate, frequency))
    return np.array(all_data)
```

##### Physics / statistics definitions.

```python
def r_avg(data):
    """ Returns the running average with MC error bars.
    """
    res = []
    for k in range(len(data)):
        res.append([k, np.mean(data[:k]), np.std(data[:k])/np.sqrt(k)])
    return np.array(res)


def a_corr(data, fit=True):
    """ Returns autocorrelation values and an extracted autocorrelation time.
    """
    pass


def N0(mu, beta, nx):
    """ Free energy on the lattice for one species.
    """
    z = np.exp(mu*beta)
    n0 = 0.
    for i in range(1,nx+1):
        for j in range(1,nx+1):
            for k in range(1,nx+1):
                ee  = 0.5 * (2*np.pi*(i-nx/2.)/nx)**2
                ee += 0.5 * (2*np.pi*(j-nx/2.)/nx)**2
                ee += 0.5 * (2*np.pi*(k-nx/2.)/nx)**2
                n0 += z*np.exp(-beta*ee)/(1. + z*np.exp(-beta*ee))
    return n0


def N0_continuum(mu, beta):
    z = np.exp(mu*beta)
    kernel = lambda x: x**2*np.exp(-x**2)/(1+z*np.exp(-x**2))
    res = integrate.quad(kernel, 0, 10)
    return res[0] * np.sqrt(32./beta)*z
```

## Virial expansion.

```python
def _Q1(beta, nx):
    """ Free energy on the lattice for one species.
    """
    n0 = 0.
    for i in range(1,nx+1):
        for j in range(1,nx+1):
            for k in range(1,nx+1):
                e  = 0.5 * (2*np.pi*(i-nx/2.)/nx)**2
                e += 0.5 * (2*np.pi*(j-nx/2.)/nx)**2
                e += 0.5 * (2*np.pi*(k-nx/2.)/nx)**2
                n0 += np.exp(-beta*(e))
    return 2*n0


def _Q2(beta, nx):
    """ NOT CHECKED.
    """
    z = np.exp(mu*beta)
    n0 = 0.
    for i in range(1,nx+1):
        for j in range(1,nx+1):
            for k in range(1,nx+1):
                e  = 0.5 * (2*np.pi*(i-nx/2.)/nx)**2
                e += 0.5 * (2*np.pi*(j-nx/2.)/nx)**2
                e += 0.5 * (2*np.pi*(k-nx/2.)/nx)**2
                for ii in range(1,nx+1):
                    for jj in range(1,nx+1):
                        for kk in range(1,nx+1):
                            ee  = 0.5 * (2*np.pi*(ii-nx/2.)/nx)**2
                            ee += 0.5 * (2*np.pi*(jj-nx/2.)/nx)**2
                            ee += 0.5 * (2*np.pi*(kk-nx/2.)/nx)**2
                            if not (ii-i) and not (jj-j) and not(kk-k):
                                n0 += np.exp(-beta*(e+ee))

# Virial coefficients. 
b2 = 1./np.sqrt(2.)
b3 = -0.3551
b31 = 0.0212
b22 = -0.0115


def ve_density(mu, h, beta, nx=9, order=3):
    """ Virial expansion for the density of the polarized Fermi gas. 
    """
    zu, zd = np.exp(beta*(mu+h)), np.exp(beta*(mu-h))
    nu, nd, nbar = N0(mu+h, beta, nx), N0(mu-h, beta, nx), 2*N0(mu, beta, nx)
    
    q1 = _Q1(beta, nx)
    res = nu + nd
    if order == 3:
        res += q1*(2*b2*zu*zd + 1.5*b3*(zu**2*zd + zu*zd**2))
    elif order == 4:
        res += 4*q1*(b22*zu**2*zd**2 + b31*(zu**3*zd + zu*zd**3)) 
    return res / nbar

def ve_magnetisation(mu, h, beta, nx=9, order=3):
    """ Virial expansion for the density of the polarized Fermi gas. 
    """
    zu, zd = np.exp(beta*(mu+h)), np.exp(beta*(mu-h))
    nu, nd, nbar = N0(mu+h, beta, nx), N0(mu-h, beta, nx), 2*N0(mu, beta, nx)
    
    q1 = _Q1(beta, nx)
    res = (nu-nd)
    if order == 3:
        res += q1*b3*(zu**2*zd - zu*zd**2)
    if order == 4:
        res += q1*2*b31*(zu**3*zd - zu*zd**3)
    return res / nbar
```

```python
param = {
    'xi' : 0.1,
    'nx': 9,
    'beta': 8.0
}
seeds = [4312, 5432, 6543]
mus = np.arange(-2.0, 3.1, 0.25)

with plt.style.context('qmstyle_slides'):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    lines = []
    
    # DLWM (DHMC, 2012) data.
    line = ax.errorbar(dmwl_data['mu'].values, dmwl_data['nn0'].values, yerr=dmwl_data['sig'], label='DHMC',
            linestyle='', marker='d', color='#FA9F42', fillstyle='full', markersize=8.)
    lines.append(line)
    
    # BDMC data.
    line = ax.errorbar(bdmc_data['betamu'].values, bdmc_data['nn0'].values, yerr=bdmc_data['sig'].values, label='BDMC',
            linestyle='', marker='s', color='#0d153a', markersize=6, fillstyle='full')
    lines.append(line)
    
    # Zwierlein data.
    line = ax.plot(zdata[0].values, zdata[1].values, label='experiment (MIT)',
            linestyle='', marker='d', color='#a42604', fillstyle='full', markeredgewidth=0., markersize=10.)
    lines.append(line[0])
    
    
    # Virial expansion.
    ve_mus = np.linspace(-4, -1.0, 20) / param['beta']
    ve = ve_density(ve_mus, 0., param['beta'], param['nx'])
    line = ax.plot(ve_mus*param['beta'], ve, label='3rd order virial exp.',
            linestyle='-', marker='', color='#FF6700', linewidth=4.)
    lines.append(line[0])
    
    for pol in [0.0]:
        param['h'] = pol
        p_lines = []
        for seed in seeds:
            p_line = []
            for mu in mus:
                param['mu'] = mu
                data = get_full_data(param, var='density', warmup_rate=0.1, frequency=1, seeds=[seed])
                n = data['n1_r'] + data['n2_r']
                p_line.append([mu, np.mean(n), np.std(n)/np.sqrt(len(n))])
            p_lines.append(p_line)
            x, y, sig = map(np.array, zip(*p_line))
            
        # Average over all seeds.
        full_line = []
        for k in range(len(mus)):
            n, sig = 0., 0.
            for l in range(len(seeds)):
                n +=  p_lines[l][k][1]
                sig += p_lines[l][k][2]**2
            n_n0b = n/len(seeds)
            sig_n0b = sig/np.sqrt(sig) 
            full_line.append([mus[k], n_n0b, sig_n0b])
        x, y_n0b, sig_n0b = map(np.array, zip(*full_line))
        
        
        # CL.
        lab = None if pol else 'complex Langevin'
        line = ax.plot(x, y_n0b, label=lab, zorder=10,
                                    marker='s', linestyle='', color='#187795')
        ax.fill_between(x, y_n0b+sig_n0b, y_n0b-sig_n0b, color=line[0].get_color(), alpha=alpha_level)
        lines.append(line[0])

        ax.set_xlim([-3.5,4.5])
        ax.set_ylim([0.9,4.0])
        ax.set_xlabel('$\\beta\\mu$')
        ax.set_ylabel('$n/\\bar{n}_0$')
        ax.legend(loc='upper left', fontsize=22, handles=lines)
        ax.tick_params(axis='both', which='major')
        fig.savefig('/home/lukas/Dropbox/work_travel/2018_02_Heidelberg/Talk/plots/UFG_EOS.svg')
        
        # Phase transition.
        bmu_pt = 2.52
        ax.axvline(bmu_pt, marker='', linestyle='-', color='gray', linewidth=4.)
        ax.text(bmu_pt-0.7, 1.0, 'NF', color='gray')
        ax.text(bmu_pt+0.3, 1.0, 'SF', color='gray')
        fig.savefig('/home/lukas/Dropbox/work_travel/2018_02_Heidelberg/Talk/plots/UFG_EOS_PT.svg')
        
```

### Chemical potentials vs. Temperature

```python
param = {
    'xi' : 0.1,
    'nx': 9,
    'h' : 0.
}
seeds = [4312, 5432, 6543]
mus = np.arange(-2.0, 4., 0.25)
betas = [8.0]


with plt.style.context('qmstyle_slides'):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    fig.set_size_inches(12, 8)
    
    
    # Zwierlein data.
    ax.plot(zdata_mu[0].values, zdata_mu[1].values, label='Zwierlein (exp)',
            linestyle='', marker='o', color='#d34f21', fillstyle='full', markeredgewidth=0., markersize=8.)

    for beta in betas:
        param['beta'] = beta
        p_lines = []
        for seed in seeds:
            p_line = []
            for mu in mus:
                param['mu'] = mu
                data = get_full_data(param, var='density', warmup_rate=0.1, frequency=1, seeds=[seed])
                n = data['n1_r'] + data['n2_r']
                p_line.append([mu, np.mean(n), np.std(n)/np.sqrt(len(n))])
            p_lines.append(p_line)
            
        # Average over all seeds.
        full_line = []
        for k in range(len(mus)):
            n, sig = 0., 0.
            for l in range(len(seeds)):
                n +=  p_lines[l][k][1]
                sig += p_lines[l][k][2]**2
            n /= len(seeds)
            sig = np.sqrt(sig)
            
            # Take out normalization.
            n0b = 2*N0_lattice(mus[k]/param['beta'], param['beta'], param['nx']) / param['nx']**3 
            n, sig = n*n0b, sig*n0b
            
            # Calculate Fermi energy (plus error).
            kF = (3.*np.pi**2 * n)**(1./3.)
            eF = kF**2/2.
            eF_sig = eF*2./3.* sig/np.abs(n)
            
            full_line.append([1./(param['beta']*eF), eF_sig/np.abs(eF) * 1./(param['beta']*eF),
                              mus[k]/(eF*param['beta']), eF_sig/np.abs(eF) * mus[k]*eF_sig/(param['beta']*eF)
                             ])
            
        x, sig_x, y, sig_y = map(np.array, zip(*full_line))
        line = ax.errorbar(x, y, xerr=sig_x, yerr=sig_y, label='CL ($N_x = $ {:d}, $\\beta = $ {:.1f})'.format(param['nx'], param['beta']), 
                    marker='s', linestyle='', color='#187795')
    
    
    # Bertsch parameter.
    xi = 0.376
    line = ax.axhline(xi, xmax=0.2, marker='', linestyle='--', color='gray', linewidth=2.5)
    ax.plot(0., xi, marker='s', color=line.get_color(), markersize=15, markeredgewidth=0., fillstyle='full')
    ax.text(0.01, 0.345, '$\\xi\ = $ {:.3f}'.format(xi), color=line.get_color(), fontsize=32)
    
    # T/Tc from experiment.
    tc = 0.167
    line = ax.axvline(0.15, marker='', linestyle='--', color='gray', linewidth=2.5)
    ax.text(0.05, 0.01, '{:.3f} $T_F$'.format(tc), color=line.get_color(), fontsize=32)


    #ax.set_xlim([-3.0,3.0])
    ax.set_xlim([0.,0.6])
    ax.set_ylim([0.,0.5])
    ax.set_ylabel('$\\mu / \\varepsilon_F$', fontsize=24)
    ax.set_xlabel('$T/T_F$', fontsize=24)
    ax.tick_params(axis='both', which='major', labelsize=24)
    ax.legend(loc='upper right', fontsize=24, numpoints=1)
    fig.savefig('/home/lukas/Dropbox/work_travel/2018_01_Hirschegg/Talk/plots//UFG_mu.svg', density=600)
```

```python
def smoothen(y, w=3):
    z = np.zeros(len(y))
    for k in range(0, len(y)):
        z[k] = np.mean(y[max(k-w,0):min(w+k,len(y))])
    return z
        
```
