```python
import numpy as np
import h5py as hdf
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline


# Workaround to not get error messages nowadays?
from matplotlib import rcParams
rcParams['text.usetex'] = True

colors = ['#0d153a', '#a42604', '#187795', '#808080']
```

### Data definitions.

```python
def _get_prefix(npart):
    if npart == '5-5' or npart == '6-4':
         return '1_small_polarization'
    if npart == '7-3':
         return '2_medium_polarization'
    if npart == '8-2' or npart == '9-1':
         return '3_large_polarization'

def get_data(param, var, warmup_rate=0., frequency=1):
    """ Returns the array for a path in the hdf5 file.
    """
    if var[:3] == 'det':
        data_file = '/home/lukas/Dropbox/GSPolarized1D/Data/CL/GSP1D_determinants.hdf5'
    elif var[:2] == 'e_':
        data_file = '/home/lukas/Dropbox/GSPolarized1D/Data/CL/GSP1D_energies.hdf5'
    else:
        data_file = '/home/lukas/work_local/1DGSP/GSP1D_correlators.hdf5'
        
    s = '{:s}/'.format(_get_prefix(param['npart']))
    grp = (s + 
           '/reg_{:.2g}'.format(param['reg']) + 
           '/clstep_{:.2g}'.format(param['clDT']) + 
           '/clinterval_{:.1f}'.format(param['clT']) + 
           '/gamma_{:.1f}'.format(param['gamma']) +
           '/npart_{:s}'.format(param['npart']) + 
           ('/betaef_{:.1f}'.format(param['betaef']) 
                if not (int(param['betaef'])-param['betaef']) else 
            '/betaef_{:.3g}'.format(param['betaef'])) +
           '/nx_{:d}'.format(param['nx']) +
           '/seed_{:d}/'.format(param['seed']))
    
    with hdf.File(data_file, 'r') as f:
        data = f[grp + var]
        cutoff = int(warmup_rate*len(data))
        return data[cutoff::frequency]


def get_full_data(param, var, t='scalar', warmup_rate=0., frequency=1):
    """ Retrieves a dataset across all seeds.
    """
    seeds = [1111, 2222, 3333, 4444, 5555]
    if t == 'scalar':
        all_data = []
        for seed in seeds:
            param['seed'] = seed
            all_data.extend(get_data(param, var, warmup_rate, frequency))
        return np.array(all_data)

    elif t == 'array':
        param['seed'] = seeds[0]
        res = np.array(get_data(param, var))
        for seed in seeds[1:]:
            param['seed'] = seed
            res += get_data(param, var)
        return res / len(seeds)


def part_to_pol(npart):
    """ Converts particle numbers to polarization.
    """
    p = map(float, npart.split('-'))
    return (p[0]-p[1])/(np.sum(p))

def kF(npart, L):
    return sum(map(float, npart.split('-')))/L * np.pi/2.
```

### Physical definitions.

```python
def average_and_fft(ospdm):
    """ Produces a pair distribution from the spatially distributed on-side pair density matrix.
    """
    vol = np.shape(ospdm)[0]
    
    # Average over spatial extent.
    r_dist = np.zeros(vol)
    for r in xrange(vol):
        r_dist += np.concatenate((ospdm[r,r:], ospdm[r,:r])) / vol
    
    
    # Wrap around for fft, perform fft and return result.
    res = np.fft.fftshift(np.fft.fft(r_dist)).real
    return res

def straighten(dm):
    """ From diagonal form to straight form.
    """
    l = np.shape(dm)[0]
    res = np.zeros_like(dm)
    for r in xrange(l):
        res[r, :] = np.concatenate((dm[r,r:], dm[r,:r]))
    return res
```

## Equation of state.

```python
def ba_strong(gamma):
    return -3./np.pi**2*gamma**2 + (gamma/(1.-2*gamma))**2 * (1.+4*np.pi**2/(15.*(1.-2.*gamma)**3))

def ba_weak(gamma):
    return 1. - 6.*gamma/np.pi**2 - (gamma/np.pi)**2
```

```python
param = {
    'clT' : 1.0,
    'reg' : 0.1,
    'nx' : 41,
    'betaef' : 2.5,
    'clDT' : 0.01
}

nparts = [['5-5', '1_small_polarization'],
          ['6-4', '1_small_polarization'],
          ['7-3', '2_medium_polarization'],
          ['8-2', '3_large_polarization'],
          ['9-1', '3_large_polarization']]

nparts = [['5-5', '1_small_polarization'],
          ['7-3', '2_medium_polarization'],
          ['9-1', '3_large_polarization']]


gammas = np.arange(0.2, 3.1, 0.4)
with plt.style.context('qmstyle_slides'):
    fig = plt.figure()
    axes = {
        'e_tot' : fig.add_subplot(111)
    }
    lines = []
    for m in range(len(nparts)):
        npart, prefix = nparts[m]
        s_data = pd.read_csv('/home/lukas/Dropbox/GSPolarized1D/Data/PT/Batchelor_strong_{:s}.dat'.format(npart.replace('-', 'p')), header=None, sep='\s+')
        w_data = pd.read_csv('/home/lukas/Dropbox/GSPolarized1D/Data/PT/Batchelor_{:s}.dat'.format(npart.replace('-', 'p')), header=None, sep='\s+')

        param['npart'] = npart
        e_line = []
        for k in range(len(gammas)):
            param['gamma'] = gammas[k]
            e_values = get_full_data(param, 'e_tot', warmup_rate=0.1)[::10]
            e_line.append([gammas[k], np.mean(e_values), np.std(e_values)/np.sqrt(len(e_values))])
            
        # CL data.
        lab = None if m else 'CL'
        x, y, sig = zip(*e_line)
        line = axes['e_tot'].errorbar(x, y, yerr=sig,
                               marker='s', label=lab, color=colors[m])
        if not m:
            lines.append(line)
        
        # Perturbation theory.
        lab = None if m else 'lattice PT'
        pt_cutoff = 1.5
        x, y = -w_data[0].values, w_data[1].values
        line = axes['e_tot'].plot(x[x<pt_cutoff], y[x<pt_cutoff], 
                           linestyle='--', color=line[0].get_color(), marker='', label=lab)[0]
        if not m:
            lines.append(line)
    
    
    axes['e_tot'].set_xlabel('$|\\gamma|$')
    axes['e_tot'].set_ylabel('$E/E_{FG}$')    
    axes['e_tot'].set_xlim([np.min(gammas)-0.2, np.max(gammas)+0.2])
    axes['e_tot'].set_ylim([-2.4,4.9])
    
    axes['e_tot'].axhline(4.02, marker='', linestyle='--', color=colors[-1])
    
    axes['e_tot'].legend(loc='lower left', handles=lines)
    fig.text(0.67, 0.88, '$N_\\uparrow\\ +\\ N_\\downarrow\\ =\\ 10$')
    axes['e_tot'].tick_params(width=2.)
    
    fig.savefig('plots/1DGSP_EOS_attractive.svg')
```

## Multiple pairing peaks.
Produces 3 plots:
 - balanced, without inset
 - balanced + polarized, without inset
 - balanced + polarized, with inset

```python
param = {
    'gamma' : 2.2,
    'clT' : 1.0,
    'reg' : 0.1,
    'nx' : 41,
    'betaef' : 2.5,
    'clDT' : 0.01
}
nparts = ['5-5', '7-3', '9-1']
colors = ['#0d153a', '#a42604', '#187795']

with plt.style.context('qmstyle_slides'):
    fig, ax = plt.subplots()
    ax.tick_params(width=2.)
    ax.set_xlabel('$k/ k_F$')
    ax.set_ylabel('$n_{pair}$')
    ax.set_xlim([-0.1, 1.5*np.pi])
    ax.set_ylim([-0.5, 8.5])
    
    qline = []
    for k in xrange(len(nparts)):
        param['npart'] = nparts[k]
        ns = map(int, param['npart'].split('-'))
        data = get_full_data(param, 'ospdm', t='array')[:].real
        
        # Rescaled k-axis k/kF.
        k_axis = np.arange(-20., 20.1, 1.0) * 2.*np.pi/param['nx'] / kF(param['npart'], param['nx'])
        
        # Calculate the FFT and plot n_pair.
        n_pair = average_and_fft(data)
        n_pair = n_pair / np.sum(n_pair) * np.prod(ns)
        line = ax.plot(k_axis, n_pair,
                linestyle='-', marker='s', color=colors[k], linewidth=4.,
                label='$N = $ {:s}'.format(param['npart'].replace('-', ' + ')))
        
        # Peak position vs. expected q of the FFLO pair.
        q = 2.*np.pi/param['nx'] * (ns[0] - ns[1])  / (2.*kF(param['npart'], param['nx']))
        qline.append([q, abs(k_axis[np.argmax(n_pair)])])
        
        
        # Labels.
        ax.text(abs(k_axis[np.argmax(n_pair)]) + 0.1, 
                        n_pair[np.argmax(n_pair)] + 0.2, 
                        '$ N =\ ${:d}$\\uparrow +\ ${:d}$\\downarrow$'.format(*ns),
                        color=line[0].get_color())
    
        # Export the balanced picture.
        if not k:
            fig.text(0.81, 0.27, '$\\gamma $ = {:.1f}'.format(param['gamma']))
            fig.savefig('/home/lukas/Dropbox/work_travel/2018_01_Hirschegg/Talk/plots/1DGSP_q_plot_balanced.svg', density=600)
                           
    # Export the polarized picture.
    fig.savefig('/home/lukas/Dropbox/work_travel/2018_01_Hirschegg/Talk/plots/1DGSP_q_plot.svg', density=600)
    
    
    # Add the inset.
    axis = fig.add_axes([0.55, 0.525, 0.4, 0.4])
    axis.tick_params(width=1.5)
    
    x, y = map(np.array, zip(*qline))
    xx = np.linspace(0., 3., 100)
    axis.plot(xx, xx, linestyle='--', marker='', 
              color='gray', label='$|k_F^{\uparrow} - k_F^{\downarrow}| / k_F$')
    for k in xrange(len(nparts)):
        axis.plot(x[k], y[k], markersize=15, fillstyle='full', markeredgewidth=0, color=colors[k])
    
    axis.set_xlim([-0.1, 2.1])
    axis.set_ylim([-0.1, 2.1])
    axis.set_xlabel('$|k_F^{\uparrow} - k_F^{\downarrow}| / k_F$', fontsize=24)
    axis.set_ylabel('$k_{peak} / k_F$', fontsize=24)
    axis.legend(loc=(0.05, 0.77), fontsize=16)
    
    # Export final picture.
    fig.savefig('/home/lukas/Dropbox/work_travel/2018_01_Hirschegg/Talk/plots/1DGSP_q_plot_inset.svg')
```
