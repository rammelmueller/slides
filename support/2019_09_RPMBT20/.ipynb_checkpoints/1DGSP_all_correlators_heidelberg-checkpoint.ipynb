```python
seeds = [1111, 2222, 3333, 4444, 5555]

np_to_param = {
    0.16 : {
        0.5 : [
            [25, (3,1)], 
            [75, (9,3)], 
            [125, (15,5)]
        ]
    },
    0.2 :{
        0.5 : [
            [20, (3,1)], 
            [60, (9,3)], 
            [100, (15,5)],
            [140, (21,7)]
        ]
    },
    0.25 : {
        0.5 : [
            [48, (9,3)], 
            [80, (15,5)], 
            [112, (21,7)]
        ]
    },
    0.4 :{
        0.0 : [
            [25, (5,5)],
            [35, (7,7)],
            [45, (9,9)],
            [55, (11,11)],
            [65, (13,13)],
            [75, (15,15)],
            [85, (17,17)],
            [95, (19,19)],
            [105, (21,21)],
            [115, (23,23)],
            [125, (25,25)],
            [135, (27,27)],
        ],
        0.25 : [
            [20, (5,3)],
            [100, (25,15)],
            [140, (35,21)],
        ],
        0.5 : [
            [30, (9,3)], 
            [50, (15,5)], 
            [70, (21,7)],
            [90, (27,9)],
            [110, (33,11)],
            [130, (39,13)]
        ],
        0.75 : [
            [20, (7,1)],
            [60, (21,3)],
            [100, (35,5)],
            [140, (49,7)],
        ]
    },
    0.5 : {
        0.5 : [
            [24, (9,3)], 
            [40, (15,5)], 
            [56, (21,7)],
            [72, (27,9)],
            [88, (33,11)],
            [104, (39,13)],
            [120, (45,15)],
            [136, (51,17)]
        ]
    },
}

def _betas(density):
    """ Get the available betas for a specific density.
    """
    return np.arange(13.0, 17.1, 2.0) if density < 0.4 else np.arange(4.0, 6.1, 1.0)

def param_to_p(param):
    return (param['npart'][0]-param['npart'][1])/(np.sum(param['npart']))

def param_to_n(param):
    return sum(param['npart'])/param['nx']
```

```python
def _get_prefix(param):
    density = param_to_n(param)
    pol = param_to_p(param)
    if density == 0.4 and pol != 0.5:
        if param['nx'] < 60:
            return '7_constant_density/71_low_volume'
        elif param['nx'] >= 110 or (param['nx']==100 and pol==0.75):
            return '7_constant_density/73_high_volume'
        return '7_constant_density/72_med_volume'
    else:   
        if param['nx'] <= 60:
            if density < 0.4:
                return '6_constant_polarization/61_low_volume'
            return '6_constant_polarization/64_low_volume_low_beta'
        elif param['nx'] >=110:
            return '6_constant_polarization/63_high_volume'
        else:
            if density < 0.4:
                return '6_constant_polarization/62_med_volume'
            return '6_constant_polarization/65_med_volume_low_beta'
        

def get_data(param, var):
    """ Returns the array for a path in the hdf5 file.
    """
    data_file = '/home/lukas/work_local/1DGSP/1D_all_correlators.hdf5'
    s = _get_prefix(param)
    grp = ('{:s}'.format(s) + 
           '/nx_{:d}'.format(param['nx']) +
           '/coupling_{:.1f}'.format(param['gamma']) +
           '/seed_{:d}'.format(param['seed']) +
           '/beta_{:.1f}'.format(param['beta']) +
           '/n_part_{:d}-{:d}'.format(*param['npart']) +
           '/mass_{:s}/'.format(param['mass'])
    )
    with hdf.File(data_file, 'r') as f:
        data = f[grp + var]
        return data[:]


def get_full_data(param, var):
    """ Retrieves a dataset across all seeds.
    """
    param['seed'] = seeds[0]
    res = np.array(get_data(param, var))
    for seed in seeds[1:]:
        param['seed'] = seed
        res += get_data(param, var)
    return res / len(seeds)
```

```python
def average_and_fft(ospdm):
    """ Produces a pair distribution from the spatially distributed on-side pair density matrix.
    """
    vol = np.shape(ospdm)[0]
    
    # Average over spatial extent.
    r_dist = np.zeros(vol)
    for r in range(vol):
        r_dist += np.concatenate((ospdm[r,r:], ospdm[r,:r])) / vol
    
    # Wrap around for fft, perform fft and return result.
    res = np.fft.fftshift(np.fft.fft(r_dist)).real
    return res


def straighten(dm):
    """ From diagonal form to straight form.
    """
    l = np.shape(dm)[0]
    res = np.zeros_like(dm)
    for r in range(l):
        res[r, :] = np.concatenate((dm[r,r:], dm[r,:r]))
    return res
```

### Handy shortcuts.

```python
def get_grid(param, norm=1.):
    """ Returns the symmetric box extent as axis for a plot, scaled with the specified norm.
    """
    grid = np.arange(-param['nx']//2, param['nx']//2) + int(param['nx'] % 2)
    if norm == 'k':
        norm = 2*np.pi/param['nx']
    return norm * grid


def wavenumbers(param, bare=False):
    """ Returns the important wavevectors:
            
            kF_up - kF_down, kF_up + kF_down
    """
    kF_up, kF_down = np.array(param['npart']) / param['nx']*np.pi
    
    q = int(np.round((kF_up-kF_down)*param['nx']/(2*np.pi)))
    l = int(np.round((kF_up+kF_down)*param['nx']/(2*np.pi)))
    
    if bare:
        return q, l
    return q*2*np.pi/param['nx'], l*2*np.pi/param['nx']
```

## Plot a single dataset.
Compare different seeds and compute the average.

```python
param = {
    'npart' : [51, 17],
    'nx' : 136,
    'gamma' : 3.0,
    'beta' : 4.0,
    'mass' : '1.0-1.0'
}
k_axis = get_grid(param, 'k')
n2k_avg = np.zeros(len(k_axis))

with plt.style.context('qmstyle_notebook'):
    fig, ax = plt.subplots()
    
    for seed in seeds:
        param['seed'] = seed
        
        ospdm = get_data(param, 'ospdm')[:].real
        n2k = average_and_fft(ospdm)
        n2k_avg += n2k/len(seeds)
        ax.plot(k_axis, n2k, linestyle='-', marker='s', alpha=0.4)
    
    ax.plot(k_axis, n2k_avg, linestyle='-', marker='', color='black')
    
    ax.set_xlabel('$k$')
    ax.set_ylabel('$n_{pair}$')
```

## Detect FFLO pattern.

```python
def op_fflo(x, A, Q,  delta):
    return A*np.abs(np.cos(Q*x)) / (x**delta)
```

## Exact situations.

### Non-interacting case $\gamma=0$

```python
def n2k_free(param, normed=True):
    """ Solution at zero interaction.
    
        Somewhat dirty solution, as it is only built by hand knowing how the result should look like. This 
        could (should) be done by other means:
        
            - either exact analytic solution (see notes from Jens)
            - convolution of two single-particle distributions
            
        In any way, the solution works.
    """
    q, l = wavenumbers(param, bare=True)

    n2k = np.zeros(param['nx'] // 2)
    n2k[0:q] = 1.
    n2k[q:l] = [1-x/(l-q) for x in range(l - q)]
    n2k = np.concatenate(([0],n2k[::-1][:-1], n2k))

    if normed:
        n2k = n2k/sum(n2k) * np.prod(param['npart']) / param['nx']
    return n2k


def rho2_free(param, normed=True):
    rho2 = np.fft.fftshift(np.fft.ifft(np.fft.fftshift(n2k_free(param, normed)))).real
    if normed:
        return rho2 / sum(np.abs(rho2))
    return rho2
```

```python
param = {
    'npart' : [9, 1],
    'nx' : 60,
}
k_axis = get_grid(param, norm='k')
r_axis = get_grid(param)

with plt.style.context('qmstyle_notebook'):
    fig = plt.figure()
    ax, ax_rho = fig.add_subplot(211), fig.add_subplot(212)

    # Infinitely attractive n2k.
    n2k = n2k_free(param)
    ax.plot(k_axis, n2k, linestyle='-', marker='', color='red')
    
    # Scale checks.
    q, l = wavenumbers(param)
    ax.axvline(q, marker='', linestyle='--', color='gray')
    ax.axvline(l, marker='', linestyle='--', color='black')
    ax.axvline(-q, marker='', linestyle='--', color='gray')
    ax.axvline(-l, marker='', linestyle='--', color='black')
    
    # 2-body denisty matrix.
    rho2_0 = rho2_free(param)
    ax_rho.plot(r_axis, np.abs(rho2_0), linestyle='-', marker='', color='red')
    
    # Expected FFLO form (without algebraic decay).
    rho2_fflo = op_fflo(r_axis, 0.0045, q, .0)
    ax_rho.plot(r_axis, rho2_fflo/sum(np.abs(rho2_fflo)),
           linestyle='--', marker='', color='gray')
    
    
    ax.set_xlabel('$k$')
    ax.set_ylabel('$n_{pair}$')
    
    ax_rho.set_xlabel('$x$')
    ax_rho.set_ylabel('$\\rho_2$')
    #ax_rho.set_xlim([-5*np.pi/q,5*np.pi/q])
        
    fig.text(0.78, 0.83, '$N = $ {:d} + {:d}'.format(*param['npart']), fontsize=24)
```

### Extreme case $\gamma\rightarrow\infty$

```python
def n2k_inf(param, normed=True):
    """ Solution at infinite attraction.
    """
    q, l = wavenumbers(param)
    n2k = np.zeros(param['nx'])
    n2k[int(param['nx']/2-q/(2*np.pi/param['nx']))] = .5
    n2k[int(param['nx']/2+q/(2*np.pi/param['nx']))] = .5
    if normed:
        n2k = n2k/sum(n2k) * np.prod(param['npart']) / param['nx']
    return n2k
   
    
def rho2_inf(param, normed=True):
    rho2 = np.fft.fftshift(np.fft.ifft(np.fft.fftshift(n2k_inf(param, normed))))
    if normed:
        return rho2 / np.sum(np.abs(rho2))
    return rho2
```

```python
param = {
    'npart' : [26, 1],
    'nx' : 60,
}

k_axis = np.arange(-param['nx']//2, param['nx']//2) * 2*np.pi/param['nx']
r_axis = np.arange(-param['nx']//2, param['nx']//2)

with plt.style.context('qmstyle_notebook'):
    fig = plt.figure()
    ax, ax_rho = fig.add_subplot(211), fig.add_subplot(212)

    # Infinitely attractive n2k.
    n2k = n2k_inf(param)
    ax.plot(k_axis, n2k, linestyle='-', marker='', color='red')
    
    # Scale checks.
    q, l = wavenumbers(param)
    ax.axvline(q, marker='', linestyle='--', color='gray')
    ax.axvline(l, marker='', linestyle='--', color='black')
    ax.axvline(-q, marker='', linestyle='--', color='gray')
    ax.axvline(-l, marker='', linestyle='--', color='black')
    
    # 2-body denisty matrix.
    rho2_0 = rho2_inf(param)
    ax_rho.plot(r_axis, np.abs(rho2_0), linestyle='-', marker='', color='red')
    
    # Expected FFLO form (without algebraic decay).
    rho2_fflo = op_fflo(r_axis, 0.0045, q, 0.)
    ax_rho.plot(r_axis, rho2_fflo/sum(np.abs(rho2_fflo)),
           linestyle='--', marker='', color='gray')
    
    
    ax.set_xlabel('$k$')
    ax.set_ylabel('$n_{pair}$')
    
    ax_rho.set_xlabel('$x$')
    ax_rho.set_ylabel('$\\rho_2$')
        
    fig.text(0.78, 0.83, '$N = $ {:d} + {:d}'.format(*param['npart']), fontsize=24)
```

### Resolution issues.

The larger $m \equiv \frac{N_\uparrow - N_\downarrow}{N_x}$, the more issues we have with resolving the oscillations. Of course, for lattices that would be powers of $2$, we'd be fine, but this would prevent fixing the density and or polarization across runs. Another way out is to use large lattices. 

We have our main runs at density $n = 0.4$ with polarizations of $p = 0, 0.25, 0.5, 0.75$ leading to $m = 0, 0.1, 0.2, 0.3$ respectively. All runs but the ones at $m=0.3$ should be fine to evaluate.

```python
from scipy.interpolate import interp1d


def plot_line(ax, param):
    r_axis = get_grid(param, norm='k')
    pbar = param_to_p(param) * param_to_n(param)
    
    rho2 = rho2_inf(param, normed=True)
    line = ax.plot(r_axis, np.abs(rho2), linestyle='-', marker='', label='$m = $ {:.3f}'.format(pbar))

    
def plot_interpolated_line(ax, param):
    r_axis = get_grid(param, norm='k')
    pbar = param_to_p(param) * param_to_n(param)
    
    rho2 = rho2_inf(param, normed=True)
    
    rho2_int = interp1d(r_axis, rho2, kind='cubic')
    r_axis_fine = np.linspace(np.min(r_axis), np.max(r_axis), 1000)
    
    line = ax.plot(r_axis_fine, np.abs(rho2_int(r_axis_fine)),
           linestyle='-', marker='', label='$m = $ {:.3f}'.format(pbar))
```

```python
# Test cases.
grid_lines = [
    {'npart': [9,1], 'nx' : 16},
    {'npart': [9,1], 'nx' : 24},
    {'npart': [9,1], 'nx' : 32},
    {'npart': [9,1], 'nx' : 64},
    {'npart': [9,1], 'nx' : 96},
    {'npart': [9,1], 'nx' : 128}
]
    
with plt.style.context('qmstyle_notebook'):
    fig = plt.figure()
    ax, ax_int = fig.add_subplot(211), fig.add_subplot(212)

    for line in grid_lines:
        plot_line(ax, line)
        plot_interpolated_line(ax_int, line)

    # Plot cosmetics.
    ax.set_xlabel('$r$')
    ax.set_ylabel('$\\rho_2$')
```
