### OBDM.

```python
from scipy.interpolate import spline, griddata


def sliding_average(l, b=1):
    """ Moves a box of length 2*b+1 along an axis and replaces the values with
        the box average (the boxes on the edges are cropped).
    """
    res = np.zeros_like(l)
    for k in range(len(l)):
        ind = np.array(range(k-b, k+b+1, 1))
        ind = ind[ind>=0]
        ind = ind[ind<len(l)]
        res[k] = np.mean(np.take(l, ind))
    return res


```

```python
etas = [3.0, 2.0, 1.0, 0.5, 0.0, -0.5, -1.0, -1.5]
npart, nx = 36, 32
kF = np.sqrt(2.*np.pi*npart/nx**2)

with plt.style.context('qmstyle_slides'):
    fig = plt.figure()
    ax = fig.add_subplot(111)

    # Set some properties.
    prop = {
        3.0 : {'axes': [ax], 'col' : '#2274a5', 'box' : 0, 'dtau' : '005', 'beta' : 10.0},
        2.0 : {'axes': [ax], 'col' : '#5f9049', 'box' : 1, 'dtau' : '005', 'beta' : 10.0},
        1.0 : {'axes': [ax], 'col' : '#9ca903', 'box' : 1, 'dtau' : '005', 'beta' : 10.0},
        0.5 : {'axes': [ax], 'col' : '#d7b501', 'box' : 2, 'dtau' : '005', 'beta' : 10.0},
        0.0 : {'axes': [ax], 'col' : '#eca300', 'box' : 3, 'dtau' : '001', 'beta' : 10.0},
        -0.5 : {'axes': [ax], 'col' : '#db7502', 'box' : 3, 'dtau' : '001', 'beta' : 6.0},
        -1.0 : {'axes': [ax], 'col' : '#cc5101', 'box' : 3, 'dtau' : '001', 'beta' : 3.0},
        -1.5 : {'axes': [ax], 'col' : '#bf3100', 'box' : 3, 'dtau' : '001', 'beta' : 2.0},
    }
    
    #free line.
    data= pd.read_csv(data_dir + "g1_free_2d.dat", sep='\s+', header=None)
    x, y = data[0].values, data[1].values
    line = ax.plot(x, y, marker='', color='gray', alpha=0.8, linestyle='--', label='$\\eta = \\infty$')
    handles = [line[0]]

    for eta in etas:
        data = pd.read_csv(data_dir +
            'greensFunctions_AVG/dt_{:s}/2D_GF_N{:d}_Nx{:d}_eta{:.1f}_beta{:.1f}.dat'
            .format(prop[eta]['dtau'], npart, nx, eta, prop[eta]['beta']), sep='\s+', header=None)

        x, y, sig = map(np.array, [np.sqrt(data[0])*kF, data[1], data[2]])
        x_smooth = np.linspace(np.min(x), np.max(x), 300)
        y_smooth = griddata(x, y, (x_smooth), method='linear')
        for ax in prop[eta]['axes']:
            line = ax.errorbar(x[sig<1.], y[sig<1.]/y[0], yerr=sig[sig<1.]/y[0],
                linestyle='', marker='s', color=prop[eta]['col'], linewidth=2.5,
                label='$\\eta = $ {:.1f}'.format(eta))
            ax.plot(x_smooth, sliding_average(y_smooth, b=1)/y[0],
                linestyle='-', marker='', color=prop[eta]['col'], linewidth=2.5,
                label='$\\eta = $ {:.1f}'.format(eta))[0]
        handles.append(line)


    # plot cosmetics.
    ax.set_xlim([0.0,7.5])
    ax.set_ylim([-0.2, 1.1])

    ax.set_xlabel('$k_F\\ r$')
    ax.set_ylabel('$\\rho^{(1)}(r)$')
    #ax.legend(loc='upper right', handles=handles, ncol=2)
    fig.text(0.68, 0.88, '$V = $ 32 $\\times$ 32')
    fig.text(0.68, 0.81, '$N = $ 18$\\uparrow$ + 18$\\downarrow$')

    #fig.tight_layout()
    fig.savefig(out_dir + '2D_obdm.svg', density=600)
```

## HMC at finite temperature (Drut, Anderson '15)

```python
from scipy.integrate import quad

def calc_b2(beta_eb):
    b2 = -0.25 + np.exp(beta_eb)
    kernel = lambda x: 2*np.exp(-beta_eb*x**2) / (x*(np.pi**2 + 4*np.log(x)**2))
    b2 -= quad(kernel, 0, np.inf)[0]
    return b2
               
def ve_2nd_order(x, b2):
    return x + 2*b2*x**2

def n0(betamu):
    z = np.exp(betamu)
    kernel = lambda x: x*(np.exp(x**2)/z+1)**(-1) 
    return 2*quad(kernel, 0, np.inf)[0]
```
